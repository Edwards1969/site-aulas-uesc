__Exercício 1__ – Balanço de Massa em Correntes

Uma planta de separação de petróleo possui três correntes de entrada e duas de saída. As composições molares (em frações) dos componentes **A**, **B** e **C** estão dadas pelas matrizes abaixo:

\begin{equation}
E =
\begin{bmatrix}
0.5 & 0.3 & 0.2 \\
0.4 & 0.4 & 0.2 \\
0.2 & 0.5 & 0.3
\end{bmatrix}
\end{equation}

\begin{equation}
S =
\begin{bmatrix}
0.45 & 0.35 & 0.20 \\
0.25 & 0.50 & 0.25
\end{bmatrix}
\end{equation}

**Pergunta:**  
Calcule a matriz diferença 

$$D = E - S$$


e interprete o significado físico dessa diferença no contexto de balanço de massa.

__Obs__: Declare todas as variáveis com $a_{ik}$

__Exercício 2 – Eficiência de um Trocador de Calor__

Um trocador de calor tem eficiência térmica representada pela multiplicação de matrizes.  A matriz de transferência de calor é dada por:

\begin{equation}
T = 
\begin{bmatrix}
0.8 & 0.1 \\
0.2 & 0.9
\end{bmatrix}
\end{equation}

E os fluxos de calor de entrada (em kW) são:

\begin{equation}
Q_{in} =
\begin{bmatrix}
200 \\
150
\end{bmatrix}
\end{equation}

__Pergunta:__ 
Determine os fluxos de saída 

$$
Q_{out} = T \cdot Q_{in}
$$


__Exercício 3 – Combinação de Reatores__

Dois reatores em série produzem três componentes químicos. As taxas de conversão nos reatores são dadas pelas matrizes:

\begin{equation}
R_1 =
\begin{bmatrix}
0.6 & 0.2 & 0.2 \\
0.1 & 0.7 & 0.2 \\
0.0 & 0.3 & 0.7
\end{bmatrix}
\end{equation}

\begin{equation}
R_2 =
\begin{bmatrix}
0.7 & 0.2 & 0.1 \\
0.2 & 0.6 & 0.2 \\
0.1 & 0.3 & 0.6
\end{bmatrix}
\end{equation}

__Pergunta:__ 
Determine a matriz global de conversão:  

$$
R = R_2 \times R_1
$$

__Exercício 4 – Mistura de Correntes__

Três correntes de entrada têm as seguintes vazões (em kg/h):

\begin{equation}
C_1 =
\begin{bmatrix}
100 \\
200 \\
150
\end{bmatrix},
\quad
C_2 =
\begin{bmatrix}
120 \\
180 \\
160
\end{bmatrix}
\end{equation}

A mistura total é:

$$M = C_1 + C_2$$


__Pergunta:__
Calcule a matriz $M$ e interprete o resultado.

__Exercício 5 – Normalização de Dados Experimentais__

Em um laboratório, os valores medidos de concentração de 2 compostos em 2 amostras foram:

\begin{equation}
C =
\begin{bmatrix}
8 & 4 \\
6 & 2
\end{bmatrix}
\end{equation}

Para normalizar os dados, deseja-se dividir toda a matriz pelo valor máximo encontrado.

__Pergunta:__ 
Calcule a matriz normalizada:

$$
C_{norm} = \frac{C}{\max(C)}
$$
