<div style="background-color: #e6f2ff; border-left: 6px solid #0074B7; padding: 10px; font-size: 16px;">
  <strong>Processo de Destilação Multicomponentees.</strong> 
</div>

Suponha que um sistema de separação de Petróleo com as seguintes entradas e saídas.

- Entrada (alimentação): Composição dos componentes: $C_1, C_2, ...,C_n$.
- Saía (produto): Produtos destilados leves, intermediários e pesados.

Para simplificar, vamos considerar três frações principais:

- Leve(L);
- Intermediário (I);
- Pesado (H)

O objetivo é estabelecer um balanço de massa e calcular as frações de cada componente nos produtos finais usando métodos matriciais.

__2.1 - Matrizes de Composição da Alimentação__

Se a composição de entrada da corrente de Petróleo bruto é represenado por um vetor de concentração:

\begin{equation}
x =
    \begin{bmatrix}
        x_1 \\
        x_2 \\
        \vdots \\
        x_3
    \end{bmatrix}
\end{equation}

onde, $x_i$ é a fração molar do componente $i$ na alimentação.

__2.2 - Matriz de Separação__

Considere que cada componente tem uma fração para cada fração de produto (leve, intermediário e pesado), criamos uma matriz de separação $S$ que representa o porcentual de cada componente em cada produto. 

\begin{equation}
S = 
\begin{bmatrix}
S_{L1} & S_{L2} & S_{Ln} \\
S_{I1} & S_{I2} & S_{In} \\
S_{H1} & S_{H2} & S_{Hn} \\
\end{bmatrix}
\end{equation}

onde:

- $S_{L1}$ representa a fração do componente $C_1$ que vai para a fração leve;
- $S_{I1}$ representa a fração do componente $C_2$ que vai para a fração Intermediçária;
- $S_{H1}$ representa a fração do componente $C_3$ que vai para a fração pesada.

E assim por diante.

__3 - Cálculo das frações dos produtos.__

Para calcular a fração de cada componente em cada fração de produto, usamos a multiplicação matricial.

$$ y = S \times x \,\,\ \text{eq.1}$$

onde:

\begin{equation}
y = 
    \begin{bmatrix}
    y_L \\
    y_I \\
    y_H
    \end{bmatrix}
\end{equation}

onde:

- $y_L$ é a fração total dos componentes na fração leve;
- $y_I$ na fração Intermediária;
- $y_H$ na fração Pesada.

\begin{equation}
S = 
    \begin{bmatrix}
        S_{L1} & S_{L2} & S_{Ln} \\
        S_{I1} & S_{I2} & S_{In} \\
        S_{H1} & S_{H2} & S_{Hn} \\
    \end{bmatrix}
\end{equation}

essas frações vão para cada fração do produto.


__2.4 - Exemplo numérico simplificado__

Suponha que temos três componentes principais ($n=3$) com frações de entrada.

\begin{equation}
S = 
    \begin{bmatrix}
        S_{L1} & S_{L2} & S_{Ln} \\
        S_{I1} & S_{I2} & S_{In} \\
        S_{H1} & S_{H2} & S_{Hn} \\
    \end{bmatrix}
    = 
    \begin{bmatrix}
        0.9 & 0.1 & 0.0 \\
        0.1 & 0.8 &0.1 \\
        0.0 & 0.1 & 0.9 \\
    \end{bmatrix}
\end{equation}

\begin{equation}
y = 
    \begin{bmatrix}
        0.9 & 0.1 & 0.0 \\
        0.1 & 0.8 &0.1 \\
        0.0 & 0.1 & 0.9 \\
    \end{bmatrix}
    = 
    \begin{bmatrix}
        0.4 \\
        0.35 \\
        0.25 \\
    \end{bmatrix}  
\end{equation}

Isso resulta nas frações dos componentes em cada uma das frações de saída (leve, intermediária e pesada).

Essa abordagem pode ser ampliada para incluir mais componentes, mais frações de forma a considerar mais variáveis adicionais como eficiência de separação, equilibrio químico, ou até acoplamento com reações químicas.

<div style="background-color: #e6f2ff; border-left: 6px solid #0074B7; padding: 10px; font-size: 16px;">
  <strong>Processo de Destilação Multicomponentees com Entradas e Saídas Múltiplas.</strong> 
</div>

Suponha um processo de destilação que separa uma corrente de alimentação de Petróleo bruto em várias frações com diferentes pontos de ebulição (com frações leves, médias e pesadas). Esse processo pode ter várias entradas, representando correntes de matéria-prima e refluxo, e varias saídas, como os produtos destilados em diferentes estágios de pureza.

__3.1 - Representação das correntes de entrada e saída com Matrizes__

Para modelar o sistema, vamos considerar:

- __Entrada__: Representado por um vetor de concentração para dado componente presente em cada corrente de entrada. Se temos $m$ corrente de entrada e $n$ componentes (hidrocarbonetos), podemos representá-los com uma matriz de entrada $E$ de dimensões $n \times m$.

\begin{equation}
E = 
\begin{bmatrix}
e_{n} & e_{12} & \cdots & e_{1m} \\
e_{21} & e_{22} & \cdots & e_{2m} \\  
\vdots & \vdots & \ddots & \vdots  \\ 
e_{n1} & e_{n2} & \cdots & e_{n3} \\
\end{bmatrix}
\end{equation}

onde  $s_{ij}$ é a concentração de componente $i$ na corrente de saída $j$.





__3.2 - Equação Geral para o Balanço de Massa.__

Para determinar as concentrações finais de cada componente nas correntes de saída utilizamos uma multiplicação matricial entre a matriz de transferência e a matriz de entrada.

$$ S = T \times E $$

ou expandindo:

\begin{equation}
    \begin{bmatrix}
        s_{11} & s_{12} & \cdots & s_{1p} \\
        s_{21} & s_{22} & \cdots & s_{2p} \\
        \vdots & \vdots & \ddots & \vdots  \\ 
        s_{n1} & s_{n2} & \cdots & s_{np} \\
    \end{bmatrix}
    = 
    \begin{bmatrix}
        t_{11} & t_{12} & \cdots & t_{1m} \\
        t_{21} & t_{22} & \cdots & t_{2m} \\
        \vdots & \vdots & \ddots & \vdots  \\ 
        t_{n1} & t_{n2} & \cdots & t{nm} \\
    \end{bmatrix}
    \times
    \begin{bmatrix}
        e_{11} & e_{12} & \cdots & e_{1m} \\
        e_{21} & e_{22} & \cdots & e_{2m} \\  
        \vdots & \vdots & \ddots & \vdots  \\ 
        e_{n1} & e_{n2} & \cdots & e_{nm} \\
\end{bmatrix}
\end{equation}

Cada produto destilado será obtido pela combinação das frações de entrada definidas pela matriz $T$ multiplicada pelas concentrações dos componentes em $E$.

__3.3 - Exemplo numérico simplificado.__

Imagine um sistema com três componentes principais e duas correntes de entrada e saída. A matriz pode ser:


- Matriz de entrada $E$:

\begin{equation}
E = 
    \begin{bmatrix}
        0.5 & 0.4 \\
        0.3 & 0.2 \\
        0.2 & 0.4 
    \end{bmatrix}
\end{equation}

- Matriz de transferência $T$:

\begin{equation}
T = 
    \begin{bmatrix}
        0.8 & 0.2 \\
        0.2 & 0.3
    \end{bmatrix}
\end{equation}

Aplicando $S = T \times E$, obtemos a matriz de saída que indica a concentração dos componentes em cada produto.  Esse método permite expandir para um número arbitrário de correntes de entrada e saída, sendo fundamental para otimizar processo, identificar gargalos e distribuir de maneira eficiente a separação dos componentes em complexas operações petroquímicas. 

Quando um processo de separação tem multiplas saídas, as matrizes de entrada e saída precisam ser ajustadas para refletir corretamente a distribuição de componentes nas diferentes frações de produtos.

Abaixo esta uma maneira de organizar essas matrizes.

__4 - Matriz de Entrada E__

A matriz de entrada continua a ser uma matriz que representa as concentrações dos componentes na corrente de entrada. Se temos $m$ correntes de entrada e $n$ componentes, a matriz de entrada será:

\begin{equation}
E = 
    \begin{bmatrix}
        e_{11} & e_{12} & \cdots & e_{1m} \\
        e_{21} & e_{22} & \cdots & e_{2m} \\  
        \vdots & \vdots & \ddots & \vdots  \\ 
        e_{n1} & e_{n2} & \cdots & e_{nm} \\
    \end{bmatrix}
\end{equation}

onde $e_{ij}$ é a concentração do componente $i$ na corrente $j$.

5 - Matrizes de Transferência $T$.

A matriz de transferência deve agora refletir como cada compnente é distribuido entre as várias saídas. Se cada componente é distribuido etre as várias saídas. Se houver $p$ saídas (por exemplo: frações leves, intermediárias e pesadas), a matriz de transferência terá dimensões $p \times m$:

\begin{equation}
    T
    = 
    \begin{bmatrix}
        t_{11} & t_{12} & \cdots & t_{1p} \\
        t_{21} & t_{22} & \cdots & t_{2p} \\
        \vdots & \vdots & \ddots & \vdots  \\ 
        t_{n1} & t_{n2} & \cdots & t{np} \\
    \end{bmatrix}
\end{equation}

onde $t_{ij}$ é a fração do componente $k$ que vai para a saída $j$. Cada linha representa uma saída e cada coluna representa uma corrente de entrada. 

__6 - Matriz de saída $S$.__

A matrz de saída $S$ terá agora dimensões $n \times p$, representando as concentrações de cada componente em cada saída. Assim, a mariz $S$ será:

\begin{equation}
S = 
    \begin{bmatrix}
        s_{11} & s_{12} & \cdots & s_{1p} \\
        s_{21} & s_{22} & \cdots & s_{2p} \\
        \vdots & \vdots & \ddots & \vdots  \\ 
        s_{n1} & s_{n2} & \cdots & s_{np} \\
    \end{bmatrix}
\end{equation}

onde $s_{ij}$ é a concentração do componente $i$ na saída $j$. 

__7 - Balanço de Massa.__

O balanço de massa para determinar as concentrações finais nas saídas pode ser expressa como:

$$ S = T \times E $$

onde cada elemento da matriz de saída é obtida pela combinação das frações dos componentes nas correntes de entrada, multiplicada pelos elementos da matriz de transferência.



__7.1 - Exemplo numérico com multiplas saídas.__

Suponha que temos duas correntes de entrada e três componentes e queremos separá-los em três saídas. 

As matrizes poderiam ser:

- Matriz de Entrada $E$:

\begin{equation}
E =
    \begin{bmatrix}
    e_{11} & e_{12}  \\
    e_{21} & e_{22}  \\
    e_{31} & e_{32}  \\
    \end{bmatrix}
    =
    \begin{bmatrix}
     0.5 & 0.4 \\
    0.3 & 0.2 \\
    0.2 & 0.4
    \end{bmatrix}
\end{equation}

- Matriz de Transferência $T$.

\begin{equation}
T = 
    \begin{bmatrix}
    t_{11} & t_{12}  \\
    t_{21} & t_{22}  \\
    t_{31} & t_{32}  \\
    \end{bmatrix}
    =
    \begin{bmatrix}
    0.7 & 0.3 \\
    0.1 & 0.8 \\
    0.2 & 0.5
    \end{bmatrix}
\end{equation}



__7.2 - Cálculo da Matriz de saída $S$.__

Agora, ao multiplicar a matriz de transferência pela matriz de entrada, teremos:

$$ S = T \times E $$

\begin{equation}
S =
    \begin{bmatrix}
    0.7 & 0.3 \\
    0.1 & 0.8 \\
    0.2 & 0.5
    \end{bmatrix}
    \times
    \begin{bmatrix}
     0.5 & 0.4 \\
    0.3 & 0.2 \\
    0.2 & 0.4
    \end{bmatrix}
\end{equation}

Multiplicando resultará em:

\begin{equation}
S =
    \begin{bmatrix}
    0.7 \cdot 0.5 + 0.3 \cdot 0.3 & 0.7 \cdot 0.4 + 0.3 \cdot 0.2 \\
    0.1 \cdot 0.5 + 0.8 \cdot 0.3 & 0.1 \cdot 0.4 + 0.8 \cdot 0.2 \\
    0.2 \cdot 0.5 + 0.5 \cdot 0.3 & 0.2 \cdot 0.4 + 0.5 \cdot 0.2 \\
    \end{bmatrix}
\end{equation}

Calculado cada elemento, obtemos a matriz $S$ com as concentrações dos componentes nas diferentes saídas. 

Este modelo matricial é extremamente útil em proccesso de separação com múltiplas entradas e saídas, permitindo uma análise clara e sistemática das concentrações dos componentes em cada fração de produto. Ela pode ser expandida e adaptada para incluir mais componentes, conforme necessário.
