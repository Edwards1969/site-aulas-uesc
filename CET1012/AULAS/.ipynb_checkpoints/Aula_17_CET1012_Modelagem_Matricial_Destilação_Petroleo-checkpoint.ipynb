Em uma unidade de __destilação atmosférica de petróleo bruto__, o engenheiro químico deseja modelar as frações de componentes em diferentes correntes de saída. A corrente de alimentação contém __seis componentes principais__ do petróleo:


- $C_1$ – Metano ($CH_4$)
- $C_2$ – Etano ($C_2H_6$)
- $C_3$ – Propano ($C_3H_8$)
- $C_4$ – Butano ($C_4H_{10}$)
- $C_5$ – Pentano ($C_5H_{12}$)
- $C_6$ – Hexano ($C_6H_{14}$)
	
	
O processo possui duas correntes de entrada:


- __Corrente 1__ — alimentação principal de petróleo bruto.
- __Corrente 2__ — corrente de refluxo contendo frações mais leves.


E três correntes de saída:

- __Saída 1__ — Fração leve (GLP e gases leves);
- __Saída 2__ — Fração intermediária (nafta leve e pesada);
- __Saída 3__ — Fração pesada (óleo combustível).

As composições das correntes de entrada são representadas pela matriz $E$ ($6 \times 2$), em que cada linha indica a **fração molar de um componente** (de $C_1$ até $C_6$) nas duas correntes de entrada do sistema.  
A **Equação (1)** apresenta essas composições:

$$
\begin{equation}
E = 
\begin{bmatrix}
	0.40 & 0.55 \\
	0.25 & 0.15 \\
	0.15 & 0.10 \\
	0.10 & 0.08 \\
	0.07 & 0.07 \\
	0.03 & 0.05
\end{bmatrix}
\end{equation}
\tag{1}
$$

Cada __coluna__ representa uma corrente de entrada, enquanto as __linhas__ indicam a participação molar de cada componente. Por exemplo, o componente $C_1$ possui frações molares de 0.40 e 0.55 nas correntes 1 e 2, respectivamente.

A __Equação (2)__ define a matriz de transferência $T$ ($3 \times 2$), na qual cada elemento $t_{ij}$ representa a __fração da corrente de entrada__ $j$ que contribui para a __saída__ $i$:


$$
\begin{equation}
T = 
\begin{bmatrix}
	0.7 & 0.3 \\
	0.2 & 0.5 \\
	0.1 & 0.2
\end{bmatrix}
\end{equation}
\tag{2}
$$


__Problemas a serem resolvidos__

a -  Calcular manualmente a matriz de saída $S = T \times E$, apresentando todas as etapas do cálculo matricial.

b - Implementar o mesmo cálculo em Python, declarando claramente todas as variáveis, utilizando a biblioteca __NumPy__.

c - Gerar um gráfico de barras empilhadas que mostre a distribuição dos seis componentes em cada corrente de saída.

d - Comentar, com base nos resultados obtidos, se as frações mais leves ($C_1–C_3$) se concentram nas saídas esperadas.

__Resolução Manual__

Sabemos que:
$$
S = T \times E
$$

Efetuando a multiplicação:

$$
S =
\begin{bmatrix}
0.7 & 0.3 \\
0.2 & 0.5 \\
0.1 & 0.2
\end{bmatrix}
\times
\begin{bmatrix}
0.40 & 0.55 \\
0.25 & 0.15 \\
0.15 & 0.10 \\
0.10 & 0.08 \\
0.07 & 0.07 \\
0.03 & 0.05
\end{bmatrix}
$$

Como as dimensões de $T$ são $3 \times 2$ e de $E$ são $6 \times 2$, precisamos transpor $E$ para compatibilizar a operação:

$$
S = T \times E^T
$$

Logo:

$$
E^T = 
\begin{bmatrix}
0.40 & 0.25 & 0.15 & 0.10 & 0.07 & 0.03 \\
0.55 & 0.15 & 0.10 & 0.08 & 0.07 & 0.05
\end{bmatrix}
$$

Multiplicando termo a termo:


$$
S =
\begin{bmatrix}
0.7 & 0.3 \\
0.2 & 0.5 \\
0.1 & 0.2
\end{bmatrix}
\times
\begin{bmatrix}
0.40 & 0.25 & 0.15 & 0.10 & 0.07 & 0.03 \\
0.55 & 0.15 & 0.10 & 0.08 & 0.07 & 0.05
\end{bmatrix}
$$

$$
S =
\begin{bmatrix}
(0.7)(0.40)+(0.3)(0.55) & (0.7)(0.25)+(0.3)(0.15) & (0.7)(0.15)+(0.3)(0.10) & (0.7)(0.10)+(0.3)(0.08) & (0.7)(0.07)+(0.3)(0.07) & (0.7)(0.03)+(0.3)(0.05) \\
(0.2)(0.40)+(0.5)(0.55) & (0.2)(0.25)+(0.5)(0.15) & (0.2)(0.15)+(0.5)(0.10) & (0.2)(0.10)+(0.5)(0.08) & (0.2)(0.07)+(0.5)(0.07) & (0.2)(0.03)+(0.5)(0.05) \\
(0.1)(0.40)+(0.2)(0.55) & (0.1)(0.25)+(0.2)(0.15) & (0.1)(0.15)+(0.2)(0.10) & (0.1)(0.10)+(0.2)(0.08) & (0.1)(0.07)+(0.2)(0.07) & (0.1)(0.03)+(0.2)(0.05)
\end{bmatrix}
$$

Efetuando os cálculos numéricos, temos:

$$
S =
\begin{bmatrix}
0.445 & 0.22 & 0.135 & 0.094 & 0.07 & 0.036 \\
0.375 & 0.155 & 0.095 & 0.076 & 0.065 & 0.031 \\
0.19 & 0.055 & 0.035 & 0.026 & 0.021 & 0.013
\end{bmatrix}
$$

__Resumo didático da matriz Transposta__

A multiplicação matricial entre a matriz de transferência $T$ e a matriz de entrada $E$ deve ser feita
utilizando a transposta de $E$, pois $T$ tem dimensões $3 \times 2$ e $E$ tem dimensões $6 \times 2$.
Assim, obtemos:

$$
S = T \times E^T
\tag{4}
$$

onde $S$ terá dimensões $3 \times 6$, representando as frações molares dos seis componentes
em cada uma das três correntes de saída. 

Em Python, a operação é implementada como:

```python
S = np.dot(T, E.T)

ou de forma equivalente:

S = T @ E.T

__Implementação Computacional em Python__

````python
# Exercício: Modelagem Matricial de Destilação de Petróleo

import numpy as np
import matplotlib.pyplot as plt

# Declaração das variáveis
componentes = ['CH4', 'C2H6', 'C3H8', 'C4H10', 'C5H12', 'C6H14']
entradas = ['Alimentação', 'Refluxo']
saidas = ['Frações Leves', 'Frações Intermediárias', 'Frações Pesadas']

# Matriz de entrada E (6x2)
E = np.array([
    [0.40, 0.55],
    [0.25, 0.15],
    [0.15, 0.10],
    [0.10, 0.08],
    [0.07, 0.07],
    [0.03, 0.05]
])

# Matriz de transferência T (3x2)
T = np.array([
    [0.7, 0.3],
    [0.2, 0.5],
    [0.1, 0.2]
])
````

````python
# Cálculo da matriz de saída
S = np.dot(T, E.T)

# Exibe a matriz de saída
print("Matriz de saída S =")
print(S.round(3))

# Gráfico de barras empilhadas
fig, ax = plt.subplots(figsize=(8,5))
bottom = np.zeros(S.shape[0])

for i in range(len(componentes)):
    ax.bar(saidas, S[:,i], label=componentes[i])

ax.set_title("Distribuição dos Componentes nas Correntes de Saída")
ax.set_xlabel("Correntes de Saída")
ax.set_ylabel("Fração Molar")
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
````

__Discussão dos Resultados__

Os resultados numéricos obtidos em _Python_ coincidem com os cálculos manuais, comprovando a coerência do modelo.  
Observa-se que os componentes leves ($CH_4$, $C_2H_6$, $C_3H_8$) concentram-se majoritariamente nas frações leves e intermediárias, enquanto os componentes mais pesados apresentam maior participação nas frações intermediárias e pesadas, conforme esperado fisicamente para um processo de destilação.


__Distribuição das Frações em uma Coluna de Destilação__

Em uma coluna de destilação típica, a separação das frações ocorre de acordo com o ponto de ebulição de cada componente. De modo geral, __quanto menor o ponto de ebulição__, __mais leve__ é a fração, e ela tende a subir na coluna.  Por outro lado, componentes com __maior ponto de ebulição__ são __mais pesados__ e permanecem na parte inferior.

$$\text{Coluna de Destilação:}$$

$$
\boxed{
\begin{array}{c}
\text{Topo da Coluna} \\
\Downarrow \\
\text{Frações Leves (C}_1 \text{–C}_3\text{)} \\
\text{(gases e compostos mais voláteis)}
\end{array}
}
$$

$$
\boxed{
\begin{array}{c}
\text{Seções Intermediárias} \\
\Downarrow \\
\text{Frações Intermediárias (C}_4 \text{–C}_7\text{)} \\
\text{(gasolina leve, nafta, querosene)}
\end{array}
}
$$

$$
\boxed{
\begin{array}{c}
\text{Base da Coluna} \\
\Downarrow \\
\text{Frações Pesadas (C}_8 \text{ e superiores)} \\
\text{(óleo diesel, óleo combustível, resíduos pesados)}
\end{array}
}
$$

$$\text{Resumo:}$$

$$
\begin{aligned}
\text{Frações leves:} & \quad \text{saem no topo (fase gasosa).} \\
\text{Frações intermediárias:} & \quad \text{são retiradas por bandejas laterais (correntes intermediárias).} \\
\text{Frações pesadas:} & \quad \text{saem no fundo (fase líquida).}
\end{aligned}
$$

> __Observação:__
> Em uma destilação fracionada, podem existir diversas correntes laterais, cada uma correspondendo a uma faixa de temperatura e composição específica.  
> Essa configuração permite recuperar produtos com diferentes graus de pureza ao longo da altura da coluna.
