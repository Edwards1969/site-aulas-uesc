__Enunciado__:

Uma corrente de alimentação de \( $F = 70\ \mathrm{mol\,min^{-1}}$ \) contendo quatro componentes deve ser separada por uma sequência de três colunas de destilação acopladas em série.  

Os componentes (índices) são:
1. Hexano (índice 1)
2. Octano (índice 2)
3. Decano (índice 3)
4. Naftaleno (índice 4)

__Configuração de fluxo:__
- A alimentação ($F$) entra na Coluna 1; a corrente de topo é ($D_1$) e a de fundo é ($B_1$).
- O topo da Coluna 1 ($D_1$) alimenta a Coluna 2, que gera ($D_2$) (topo) e ($B_2$) (fundo).
- O fundo da Coluna 1 ($B_1$) alimenta a Coluna 3, que gera ($D_3$) (topo) e ($_3$) (fundo).
- Os fluxos finais de produto são ($D_2$, $B_2$, $D_3$, $B_3$).


**Tabela 1 — Dados de composição (frações molares)**

| Corrente / Componente | Hexano (1) | Octano (2) | Decano (3) | Naftaleno (4) |
|:----------------------|:-----------:|:-----------:|:-----------:|:--------------:|
| Alimentação ( $z_j$) | 0.35 | 0.25 | 0.25 | 0.15 |
| Coluna 2: topo ($y_{2j}$) | 0.40 | 0.30 | 0.20 | 0.10 |
| Coluna 2: fundo ($x_{2j}$) | 0.10292 | 0.14321 | 0.39827 | 0.35560 |
| Coluna 3: topo ($y_{3j}$) | 0.54115 | 0.24517 | 0.20441 | 0.00927 |
| Coluna 3: fundo ($x_{3j}$) | 0.05 | 0.10 | 0.40 | 0.45 |


__Parte A — Sistema matricial para os produtos finais__

O balanço de massa para cada componente é:

$$
F z_j = D_2 y_{2j} + B_2 x_{2j} + D_3 y_{3j} + B_3 x_{3j}, \quad j = 1,2,3,4
$$

Essas quatro equações podem ser escritas na forma matricial:

$$
A \mathbf{v} = \mathbf{b}
$$

onde:

$$
A =
\begin{bmatrix}
y_{21} & x_{21} & y_{31} & x_{31} \\
y_{22} & x_{22} & y_{32} & x_{32} \\
y_{23} & x_{23} & y_{33} & x_{33} \\
y_{24} & x_{24} & y_{34} & x_{34}
\end{bmatrix},
\quad
\mathbf{v} =
\begin{bmatrix}
D_2 \\ B_2 \\ D_3 \\ B_3
\end{bmatrix},
\quad
\mathbf{b} =
\begin{bmatrix}
F z_1 \\ F z_2 \\ F z_3 \\ F z_4
\end{bmatrix}
$$


Com os dados da tabela:

$$
A =
\begin{bmatrix}
0.40 & 0.10292 & 0.54115 & 0.05 \\
0.30 & 0.14321 & 0.24517 & 0.10 \\
0.20 & 0.39827 & 0.20441 & 0.40 \\
0.10 & 0.35560 & 0.00927 & 0.45
\end{bmatrix},
\quad
\mathbf{b} = 70
\begin{bmatrix}
0.35 \\ 0.25 \\ 0.25 \\ 0.15
\end{bmatrix}
=
\begin{bmatrix}
24.5 \\ 17.5 \\ 17.5 \\ 10.5
\end{bmatrix}
$$

A solução é obtida por:

$$
\mathbf{v} = A^{-1} \mathbf{b}
$$


__Parte B — Cálculo de ( $D_1$), ($B_1$) e suas composições__

$$
D_1 = D_2 + B_2, \quad B_1 = D_3 + B_3
$$

As composições de topo e fundo da Coluna 1 são dadas por:

$$
y_{1j} = \frac{D_2 y_{2j} + B_2 x_{2j}}{D_1}, \quad
x_{1j} = \frac{D_3 y_{3j} + B_3 x_{3j}}{B_1}
$$

4. Modelar o transiente da concentração de Hexano em ($D_2$):

$$
C_{D2,1}(t) = C_{D2,1}^{\infty} + \bigl(C_{D2,1}^{0} - C_{D2,1}^{\infty}\bigr)e^{-t/\tau}
$$

5. Gerar gráficos para ($\tau = 2, 5, 10$) min.

__Parte C — Modelagem do Transiente da Concentração de Hexano em__ $D_2$

Até agora, o problema foi resolvido considerando o __estado estacionário__, isto é, todas as vazões e composições permanecem constantes no tempo.  
Entretanto, em um processo real de destilação, quando ocorre uma __mudança nas condições de operação__ (como variação da vazão de alimentação ou temperatura), o sistema __não atinge o novo estado estacionário instantaneamente__. Ele passa por um __regime transiente__, no qual as variáveis mudam com o tempo até atingirem o novo equilíbrio.

A evolução temporal da concentração de Hexano na corrente de topo da Coluna 2 (corrente $D_2$) pode ser representada por um modelo de __$1^a$ ordem__, dado por:

$$
C_{D2,1}(t) = C_{D2,1}^{\infty} + \bigl(C_{D2,1}^{0} - C_{D2,1}^{\infty}\bigr)e^{-t/\tau}
$$

onde:

- ( $C_{D2,1}(t)$) é a concentração (ou fração molar) de Hexano em D₂ no instante ($t$);
- ($C_{D2,1}^{0}$) é a concentração inicial (antes da perturbação);
- ($C_{D2,1}^{\infty}$) é a concentração final (no novo estado estacionário);
- ($ \tau $) é a __constante de tempo__ do sistema, que indica a rapidez da resposta (quanto menor ($\tau$), mais rápida a estabilização);
- ($ t $) é o tempo (em minutos).

__Interpretação física__

- No instante ($t = 0$), a concentração vale ($C_{D2,1}^{0}$).
- À medida que o tempo avança, o termo exponencial ( $e^{-t/\tau}$) decai, e ($C_{D2,1}(t)$) se aproxima de ($C_{D2,1}^{\infty}$).
- Após cerca de ($4\tau$), o sistema já está praticamente em regime estacionário.

__O que você deve fazer__

1. Assumir valores, por exemplo:
   - ( $C_{D2,1}^{0} = 0.25$)
   - ( $C_{D2,1}^{\infty} = 0.40$)
2. Modelar o comportamento temporal da concentração para diferentes valores de ($\tau $) (por exemplo, $2$, $5$ e $10$ minutos).
3. Gerar o gráfico da evolução de ($C_{D2,1}(t)$) e comparar a rapidez da resposta para cada caso.

---

Na próxima célula, você pode implementar o modelo acima em Python, gerando as curvas para diferentes valores de ($\tau$).

__Simulação da evolução temporal no Python__:

$
C^o_{D2,1}=0.25, \quad C^\infty_{D2,1}=0.40,
$



```python
import numpy as np
import matplotlib.pyplot as plt

C0, C_inf = 0.25, 0.40
tau_values = [2, 5, 10]  # minutos
t = np.linspace(0, 30, 200)

plt.figure(figsize=(7,5))
for tau in tau_values:
    C_t = C_inf + (C0 - C_inf) * np.exp(-t/tau)
    plt.plot(t, C_t, label=f'τ = {tau} min')

plt.xlabel('Tempo (min)')
plt.ylabel('Fração molar de Hexano em D₂')
plt.title('Resposta transiente de 1ª ordem')
plt.legend()
plt.grid(True)
plt.show()
```

Todas as vazões são positivas, garantindo consistência física.


```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Definição dos dados do problema ---
F = 70  # mol/min

# Frações molares da alimentação
z = np.array([0.35, 0.25, 0.25, 0.15])

# Coluna 2 (topo e fundo)
y2 = np.array([0.40, 0.30, 0.20, 0.10])
x2 = np.array([0.1029198, 0.14320896, 0.39827039, 0.35560085])

# Coluna 3 (topo e fundo)
y3 = np.array([0.54115045, 0.24517245, 0.20441150, 0.00926560])
x3 = np.array([0.05, 0.10, 0.40, 0.45])
```

```python
# --- 2. Montagem da matriz A e vetor b ---
A = np.array([
    [y2[0], x2[0], y3[0], x3[0]],
    [y2[1], x2[1], y3[1], x3[1]],
    [y2[2], x2[2], y3[2], x3[2]],
    [y2[3], x2[3], y3[3], x3[3]]
])

b = F * z

# --- 3. Resolução do sistema A·v = b ---
v = np.linalg.solve(A, b)
D2, B2, D3, B3 = v
print("Vazões molares [mol/min]:")
print(f"D2 = {D2:.4f}, B2 = {B2:.4f}, D3 = {D3:.4f}, B3 = {B3:.4f}")

# --- 4. Cálculo de D1 e B1 ---
D1 = D2 + B2
B1 = D3 + B3
print(f"\nD1 = {D1:.4f} mol/min")
print(f"B1 = {B1:.4f} mol/min")

# --- 5. Composições de D1 e B1 ---
y1 = (D2 * y2 + B2 * x2) / D1
x1 = (D3 * y3 + B3 * x3) / B1
```

```python
print("\nFrações molares em D1:")
print(np.round(y1, 6))
print("Soma =", y1.sum())

print("\nFrações molares em B1:")
print(np.round(x1, 6))
print("Soma =", x1.sum())
```

```python
# --- 6. Verificação da conservação ---
# Conservação total
F_total = F
saida_total = D2 + B2 + D3 + B3
print(f"\nConservação total: {saida_total:.4f} ≈ {F_total}")

# Conservação por componente
entrada = F * z
saida = D2*y2 + B2*x2 + D3*y3 + B3*x3
print("\nDiferença por componente (entrada - saída):")
print(np.round(entrada - saida, 6))
```

```python
# --- 7. Transiente da concentração de Hexano em D2 ---
C_inf = y2[0]        # valor no estado estacionário (fração molar de Hexano)
C0 = 0.2              # valor inicial hipotético
taus = [2, 5, 10]     # tempos característicos (min)

t = np.linspace(0, 20, 200)
plt.figure(figsize=(8, 5))

for tau in taus:
    C_t = C_inf + (C0 - C_inf) * np.exp(-t / tau)
    plt.plot(t, C_t, label=f"τ = {tau} min")

plt.xlabel("Tempo (min)")
plt.ylabel("Fração molar de Hexano em D₂")
plt.title("Resposta Transiente da Composição em D₂")
plt.legend()
plt.grid(True)
plt.show()
```
