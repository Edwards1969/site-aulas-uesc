__Considerações iniciais__

O projeto tem como objetivo integrar intrumentação, modelagem dinâmica e controle de processos em um único problema que pode ser resolvido __inteiramente via programação em Python__ simulando o sistema físico.

__Enunciado__ (para o aluno)

Um tanque de aço inoxidável contém $100 \ L$ de água, aquecida por uma resistência elétrica. A temperatura deve ser controlada  em $60 ^o C$.

Um __termopar tipo K__ mede a temperatura e envia o sinal para um __transmissor__ com saída em $4-20 \ nA$ proporcionalmente à faixa de $0-100 ^o C$.

O __controlador__ ajusta a pontência elétrica ($0 - 100 \%$) de uma resistência de $1 \ kW$, que aquece o tanque.

O sistema deve ser modelado computacionalmente de forma que você dever:

1. Desenvolver o __modelo dinâmico__ de temperatura da água;

2. Implementar a __instrumentação virtual__ (sensor, transmissor, atuador);

3. Projetar e simular um __controlador PID__ para manter a temperatura em $60 ^o C$;

4. Visualizar em __gráfico de resposta temporal__ a variação da temperatura, sinal de controle e erro de regime permanente.

Considere que há perdas térmicas proporcionais à diferença de temperatura com o amgiente ($25 ^o C$).


__Etapas para desenvolvimento em Python__

1. Modelagem do Processo:

$$ C_p \cdot m \dfrac{dT}{dt} = Q_{in} - hA(T -  T_a) $$

onde:

- $Q_{in} = \eta P$ -> potêncial elétrico (controlada pelo controlador);
- $hA(T - T_a)$ -> perda térmica;
- $m = 100 \ Kg$, $C_p = 4180 \ J/Kg ^oC$;
- $T_a = 25 ^oC$






__2. Instrumentação virtual__

- __Sensor__: gera ruído aleatório (simulando erro de medição);
- __Transmissor__: converte $^oC$ -> sinal de $4-20 \ mA$
- __Atuador__: convertesaída do controlador ($0-100 \%$) -> potêncial elétrico ($0 - 1000 \ W$)

__3. Controle PID__:

- Implementar equações do PID discreto;
- Ajustar ganho $K_p, T_i, T_d$;
- Analisar resposta temporal (overshoot, tempo de acomodação, erro estacionário).

__4. Visualização gráfica__:

- Plotar gráficos: $T(t) \times x$

    - Sinal de controle $x$
    - Corrente do transmissor ($mA$)

__Ferramentas para programação em Python__

- _numpy_ -> simulação numérica (Euler ou odeint)
- _matplotlib_ -> gráficos
- _scipy_ -> integração de EDO
- _tkinter_ (opcional) -> interface para ajustar parâmetros PID.

__Dados do problema__

Abaixo a relação de parâmetros, descrição e valor, respectivamente.

- $m$ - massa da água em $100 \ Kg$
- $C_p$ - Calor específico da água, $4189 \ J/Kg ^oC$
- $P_{max}$ - Potêncial máxima da resistência, $1000 \ W$
- $hA$ - Coeficiente global de perda térmica, $15 \ W/^oC$
- $T_a$ - Temperatura ambiente, $25 ^oC$
- $T_{sp}$ - Setpoint da Temperatura, $60 ^oC$
- $\eta$ - Eficiência da resistência, $0.95$ 


__Modelo matemático__

O balançco de energia no tanque é dado por:

$$
m c_p \frac{dT}{dt} = \eta P - hA (T - T_a)
$$

onde:
- \(T \): temperatura do líquidon($^oC$)
- \(P\): potênficial elétrico (W)
- \(T_a\): temperatura ambiente ($^oC$)

__Código base em Python__

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parâmetros físicos
m = 100.0            # massa de água (kg)
Cp = 4180.0          # calor específico (J/kg°C)
eta = 0.95           # eficiência
P_max = 1000.0       # potência máxima (W)
hA = 15.0            # coeficiente global de perda térmica (W/°C)
T_amb = 25.0         # temperatura ambiente (°C)
```

```python
# Parâmetros de simulação
dt = 1.0             # passo de tempo (s)
t_final = 2000       # tempo total (s)
n = int(t_final / dt)

# Sinal de referência (setpoint)
T_sp = 60.0          # °C

# Parâmetros do PID
Kp = 8.0
Ki = 0.02
Kd = 100.0

# Variáveis de controle
integral = 0.0
erro_anterior = 0.0
```

```python
# Vetores de armazenamento
tempo = np.arange(0, t_final, dt)
T = np.zeros(n)
T[0] = 25.0          # temperatura inicial
u = np.zeros(n)      # sinal de controle (0–100%)
I_mA = np.zeros(n)   # corrente do transmissor
```

```python
# Funções de instrumentação virtual
def sensor(temp_real):
    """Sensor com ruído aleatório"""
    ruido = np.random.normal(0, 0.2)
    return temp_real + ruido

def transmissor(temp_sensor):
    """Converte 0–100 °C em 4–20 mA"""
    return 4 + (16/100) * temp_sensor

def atuador(u_percent):
    """Converte 0–100% em potência (W)"""
    return (u_percent/100) * P_max
```

```python
# Simulação
for i in range(1, n):
    # Sensor
    T_sensor = sensor(T[i-1])
    
    # Erro de controle
    erro = T_sp - T_sensor
    
    # PID discreto
    integral += erro * dt
    derivada = (erro - erro_anterior) / dt
    u[i] = Kp*erro + Ki*integral + Kd*derivada
    erro_anterior = erro
    
    # Saturação (0 a 100%)
    u[i] = max(0, min(100, u[i]))
    
    # Atuador (potência elétrica)
    P = atuador(u[i])
    
    # Modelo térmico (integração de Euler)
    dTdt = (eta*P - hA*(T[i-1] - T_amb)) / (m*Cp)
    T[i] = T[i-1] + dTdt*dt
    
    # Transmissor (corrente em mA)
    I_mA[i] = transmissor(T_sensor)
```

```python
# Plot dos resultados
plt.figure(figsize=(10,6))
plt.plot(tempo, T, label='Temperatura (°C)')
plt.axhline(T_sp, color='r', linestyle='--', label='Setpoint (°C)')
plt.xlabel('Tempo (s)')
plt.ylabel('Temperatura (°C)')
plt.legend()
plt.grid(True)
plt.title('Resposta da Temperatura no Tanque')
plt.show()

plt.figure(figsize=(10,6))
plt.plot(tempo, u, label='Sinal de controle (%)')
plt.xlabel('Tempo (s)')
plt.ylabel('Controle (%)')
plt.legend()
plt.grid(True)
plt.title('Ação de Controle PID')
plt.show()

plt.figure(figsize=(10,6))
plt.plot(tempo, I_mA, label='Sinal do Transmissor (mA)')
plt.xlabel('Tempo (s)')
plt.ylabel('Corrente (mA)')
plt.legend()
plt.grid(True)
plt.title('Sinal de Corrente Enviado pelo Transmissor')
plt.show()
```
