__Enunciado do problema: Cálculo do Volume em um Tanque Cilíndrico Horizontal__

Um tanque cilíndrico de raio $R$ e comprimento $L$, esta posicionado horizontalmente (deitado sobre seu lado). Esse tanque esta sendo preenchdo com um líquido, e o nível do líquido $h$ - medido a partir do ponto mais baixo do tanque - varia com o tempo $t$, de modo que:

$$ 0 \leq h(t) \leq 2R $$

__Objetivo__:

 Deseja-se determinar uma expressão para a área da seção transversal molhada $A(t)$ - ou seja, a área da região circular abaixo da superfície do líquido - em função de $h$. Com essa expressão, o volume total de líquido no tanque em qualquer instante $t$ será dado por:

$$  V(t) = A(h(t)) \cdot L$$

__Tarefa__:

Deduza matematicamente a expressão para $A(t)$, considerando separadamente os casos em que o líquído ocupa a metada inferior do tanque ($ 0 \leq h \leq R $) e a metada superior ($R < h \leq 2R$)).

__Informações Adicionais para Resolução__:

- A seção transversal do tanque é um circulo de raio $R$, centrado na origem, dado pela equação: $ x^2 + y^2 = R^2$.

- A superfície do líquido é uma reta horizontal na posição: $y = h - R$.

- A área $A(h)$ pode ser obtida por integração da função que descreve o comprimento horizontal do circulo em cada altura $y$, ou por uso de fórmulas geométricas de segmento circular.

__Resultado Esperado__:

Ao final, você devera apresentar a expressão final de $A(t)$ nos dois intervalos de $h$, mostrar que ela é contínua em $h = R$ e verificar os valores nos extremos $h = 0$ e $h = 2R$.

__Resolução__:

__1. Configuração do problema.__

- Raio do cilindro: $R$.

- Comprimento do tanque: $L$.

- Altura do líquido medido a partir do fundo: $h$ (varia de $0$ a $2R$).

- Queremos: Área molhada $A(h)$ da seção circular, correspondente à região preenchida pelo líquido.
__2. Geometria da seção circular.__

Considere um círculo de raio $R$, centrado na origem ($0,0$):

$$  x^2 + y^2 = R^2 $$

O líquido preenche de baixo para cima até a altura de $h$ (nedida a partir do pontomais baixo do círculo).

- O fundo do tanque esta em $ y = -R $

- O topo do tanque está em $ y = +R $

A altura do líquido $h$ significa que a superfície do líquido está na linha:

$$  y = -R + h $$

ou seja, 

$$ y = h - R $$


__3. Casos possíveis__.

__Caso 1__: $0 \leq h \leq R$

O líquido esta na parte inferior do círculo.


__Caso 2__: $R < h \leq 2R$

O líquido está acima do centro do círculo. Vamos resolver primeiro para $0 \leq h \leq R$.

__4. Cálculo da área molhada para__ $0 \leq h \leq R$

A linha do círculo: $y = h - R$

No círculo $x^2 + y^2 = R^2$ a inteseção com essa linha horizontal dá:

$$ x^2 + (h - R)^2 = R^2  $$

$$ x^2 = R^2 - (h - R)^2 $$

$$x^2 = R^2 - (h^2 - 2Rh + R^2)$$

$$ x^2 = 2Rh - h^2 $$

Seja $ a = \sqrt{2Rh - h^2}$, então os pontos de interseção são ($-a, h-R$) e ($+a, h - R$).

A largura da superfície do líquido é $2a$.

__Área da região abaixo da linha__ $y = h - R$ __dentro do círculo__.

Podemos calcular como:

$A(h)... \text{área do segmento circular}$

$$ A(h) = R^2 cos^{-1}(\frac{R - h}{R}) - (R - h)\sqrt{2Rh- h^2}  $$

Vamos verificar:


O ângulo $\theta$ é medido a partir do centro: a distância vertical do centro até a superfície do líquido é $d = R - h$, (já que o centro está em $y=0$ e a superfície em $y = h - R$, então $h - R = -d. \,\ Então: d = R - h$).

Para $h \leq R, d \geq  0$.

A fórmula clássica da área de um segmento circular de altura $H$ (medida da borda até a corda) é: 

Se o segmento é menor que um semicírclo (caso $h \leq R$), a altura do segmento $H = h$.

Mas cuidado: a altura do seguimento na literatura é a distâcia da corda até a borda no mesmo lado do centro? Vamos refazer de forma clara.

__5. Definição clara do ângulo.__

Coloque o circulo no centro ($0,0$), raio $R$.

A linha do líquido: $y = c$, onde $c = h - R$.

Para $h \leq R$, temos $c \leq 0$.

A área abaixo da linha (onde está o líquido) é:

$$ A(h) = \int_{x=-a}^{x=a} \int_{y=-R}^{y=c}? $$

Melhor: A área é a integral de $y$ de $y = -R$ até $y = c$ do comprimento horizontal naquele $y$:

Comprimento horizontal em $y=2\sqrt{R^2 - y^2}$

Então: 

$$ A(h) = \int_{y=-R}^c R^2 \sqrt{R^2 - y^2}dy $$

onde: $c = h - R$

__6. Resolvendo a integral__

Sabemos que 

$$ \int \sqrt{R^2 -y^2} =  \frac{y}{2}\sqrt{R^2 - y^2}+ \frac{R^2}{s}sin^{-1}(\frac{y}{R}) + C $$

Avaliando de $y = -R$ a $a = c$:

Em $y = -R$:

$$ \sqrt{R^2 - R^2} = 0, \,\ sin^{-1}(-1) = -\frac{\pi}{2} $$

Então a primitiva em $y = -R$ vale:

$$   -\frac{R}{2} \cdot 0 + \frac{R^2}{2} \cdot (- \frac{\pi}{2}) = -\frac{\pi R^2 }{4}$$

Em $y = c$:

Primitiva:

$$  \frac{c}{2} \sqrt{R^2 - c^2} + \frac{R^2}{2}sin^{-1}(\frac{c}{R})$$

logo:

$$ 
\int_{-R}^c \sqrt{R^2 -y^2}dy = [\frac{c}{2}\sqrt{R^2-c^2}+\frac{R^2}{2}sin^{-1}(\frac{c}{R})] - (-\frac{\pi R^2}{4}) = [\frac{c}{2}\sqrt{R^2-c^2}+\frac{R^2}{2}sin^{-1}(\frac{c}{R})] + (\frac{\pi R^2}{4}) 
$$ 

Multiplicando por 2 (já que tinhamos $2x$ a integral)

$$ A(h) = c\sqrt{R^2 - c^2} + R^2sin^{-1} (\frac{c}{R}) + \frac{\pi R^2}{2}$$











__7.Substituindo__ $c=h-R$

Temos:

$$ 
R^2 - c^2 = R^2 - (h - R)^2 = 2Rh - h^2  
$$

e

$$ 
\sqrt{R^2-c^2} = \sqrt{2Rh - h^2}
$$

Também: $c = h - R$

Então:

$$  
A(h) = (h - R) \sqrt{2Rh - h^2} +  R^2 sin^{-1} (\frac{h - R}{R}) + \frac{\pi R^2}{2}
$$

Note que:

$sin^{-1}(\frac{h - R}{R}) = sin^{-1}(1 - \frac{h}{R}) $ ( pode ser uma ou outra forma)

mas 

$sin^{-1}(\frac{h - R}{R}) = -cos^{-1}(\frac{R-h}{R}) $


Vamos ver:

Sabemos que :

$$ sin^{-1}(-x) = - sin^{-1}(x) $$

Aqui:

$$ \frac{h - R}{R} = -\frac{R - h}{R} $$

Então:

$$ sin^{-1}(\frac{h-R}{R}) = -sin^{-1}(\frac{R - h}{R}) $$

Portanto:

$$A(h) = (h-R) \sqrt{2Rh - h^2} - R^2sin^{-1}(\frac{R - h}{R}) + \frac{\pi R^2}{2}$$














__8. Simplificando__

Para $0 \leq  h\leq R$, a área molhada (parte inferior) é:

$$ A(h) = R^2 cos^{-1}(\frac{R - h}{R}) - (R - h)\sqrt{2Rh - h^2}$$

Pois:

$$ cos^{-1} = R^2 cos^{-1}(\frac{R - h}{R}) - (R - h)\sqrt{2Rh - h^2}  $$

Pois:
$$ cos^{-1}(\frac{R - h}{R}) = \frac{\pi}{2} - sin^{-1}(\frac{R - h}{R}) $$

e substituindo a expressão acima, bate. (verficação rápida confere o resultado).

__9. Caso__ $R < h \leq 2R$.

Quando $h > R$, a área molhada é a área total do círculo $\pi R^2$ menos a área não molhada (parte superior vazia), que é um segmento circular de altura $2R-h$.

Pela simetria, a área do segmento superior de altura $H = 2R - h$ é:

$$A_{\text{segmento}}(H) =  R^2cos^{-1}(\frac{R - H}{R}) - (R - H)\sqrt{2RH - H^2} $$

Substituindo: $H = 2R - h$:

$$R - H = R - (2R - h) = h - R$$

$$
2RH - H^2 = 2R(2R- h) - (2R - h)^2 
$$

$$
 = 4R^2 - 2Rh - (4R^2 - 4Rh + h^2) 
$$

$$
  = 4R^2 - 2Rh - 4R^2 -h^2 
$$

$$
 = 2Rh - h^2
$$

Que corresponde ao mesmo de antes.

Então a área do segmento superior é:

$$ A_{\text{seg \ superior}} =  R^2cos^{-1}(\frac{h - R}{R}) - (h - R)\sqrt{2Rh - h^2} $$

A área molhada é a área total menos essa área:

$$ A(h) = \pi R^2 - [R^2cos^{-1}(\frac{h - R}{R}) - (h - R)\sqrt{2Rh - h^2}]  $$

$$ A(h) = \pi R^2  -  R^2cos^{-1}(\frac{h - R}{R}) + (h - R)\sqrt{2Rh - h^2} $$


__10. Resumo final__

Para um cilindro horizontal de raio $R$, comprimento $L$, altura líquida $h$ com $0 \le h \le 2R$.


$$
A(h) = 
\begin{cases}
R^2 \cos^{-1} \left(\dfrac{R - h}{R} \right) - (R - h)\sqrt{2Rh - h^2}, & 0  \le h \le R 
\\
\pi R^2 - R^2 \cos^{-1} \left(\dfrac{h - R}{R} \right) + (h - R)\sqrt{2Rh - h^2}, & R <  h \le 2R 
\end{cases}
$$


E o volume:

$$ V(h) = A(h) \cdot L $$

__11. Observações__:

- $cos^{-1}$ retorna valores em [$a, \pi$].

- Para $h = R$ (meio cheio), $A(R) = \dfrac{\pi R^2}{2}$.

- Para $h = 0$, $A(0)=0$.

- Para $h = 2R$, $A(2R) = \pi R^2$

Essa expressão permite calcular a área molhada para qualquer $h$ e, portanto o volume do tanque.

12. Código em Python com aplicação do conceito apresentado acima.

```python
import numpy as np
import matplotlib.pyplot as plt

class TanqueCilindricoHorizontal:
    """
    Classe para cálculo do volume em tanque cilíndrico horizontal
    Baseado na teoria de segmento circular
    """
    
    def __init__(self, raio, comprimento):
        """
        Inicializa o tanque cilíndrico horizontal
        
        Parâmetros:
        raio (float): Raio do cilindro (R)
        comprimento (float): Comprimento do tanque (L)
        """
        self.R = raio
        self.L = comprimento
        
    def area_molhada(self, h):
        """
        Calcula a área molhada da seção transversal para uma altura h
        
        Baseado na teoria:
        Para 0 ≤ h ≤ R: A(h) = R² cos⁻¹((R-h)/R) - (R-h)√(2Rh - h²)
        Para R < h ≤ 2R: A(h) = πR² - [R² cos⁻¹((h-R)/R) - (h-R)√(2Rh - h²)]
        
        Parâmetros:
        h (float ou array): Altura do líquido (0 a 2R)
        
        Retorna:
        float ou array: Área molhada da seção transversal
        """
        h = np.asarray(h)
        R = self.R
        
        # Inicializa array de resultados
        A = np.zeros_like(h)
        
        # CASO 1: 0 ≤ h ≤ R (líquido na metade inferior)
        # A(h) = R² cos⁻¹((R-h)/R) - (R-h)√(2Rh - h²)
        mask_inferior = (h >= 0) & (h <= R)
        if np.any(mask_inferior):
            h_inf = h[mask_inferior]
            termo_arco = R**2 * np.arccos((R - h_inf) / R)
            termo_raiz = (R - h_inf) * np.sqrt(2 * R * h_inf - h_inf**2)
            A[mask_inferior] = termo_arco - termo_raiz
        
        # CASO 2: R < h ≤ 2R (líquido na metade superior)
        # A(h) = πR² - [R² cos⁻¹((h-R)/R) - (h-R)√(2Rh - h²)]
        mask_superior = (h > R) & (h <= 2 * R)
        if np.any(mask_superior):
            h_sup = h[mask_superior]
            termo_arco_sup = R**2 * np.arccos((h_sup - R) / R)
            termo_raiz_sup = (h_sup - R) * np.sqrt(2 * R * h_sup - h_sup**2)
            area_segmento_superior = termo_arco_sup - termo_raiz_sup
            A[mask_superior] = np.pi * R**2 - area_segmento_superior
        
        # Valores fora do intervalo [0, 2R] retornam NaN
        mask_fora = (h < 0) | (h > 2 * R)
        if np.any(mask_fora):
            A[mask_fora] = np.nan
            
        return A
    
    def volume(self, h):
        """
        Calcula o volume do líquido no tanque para uma altura h
        
        Baseado na teoria: V(h) = A(h) × L
        
        Parâmetros:
        h (float ou array): Altura do líquido (0 a 2R)
        
        Retorna:
        float ou array: Volume do líquido
        """
        A = self.area_molhada(h)
        return A * self.L
    
    def plot_curvas(self, n_points=1000):
        """
        Gera gráficos da área e volume em função da altura
        """
        alturas = np.linspace(0, 2 * self.R, n_points)
        areas = self.area_molhada(alturas)
        volumes = self.volume(alturas)
        
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
        
        # Gráfico da área
        ax1.plot(alturas, areas, 'b-', linewidth=2, label='Área Molhada')
        ax1.axvline(x=self.R, color='r', linestyle='--', alpha=0.7, label='h = R (metade)')
        ax1.set_xlabel('Altura do Líquido h (m)')
        ax1.set_ylabel('Área Molhada A(h) (m²)')
        ax1.set_title('Área Molhada vs Altura')
        ax1.grid(True, alpha=0.3)
        ax1.legend()
        
        # Gráfico do volume
        ax2.plot(alturas, volumes, 'g-', linewidth=2, label='Volume')
        ax2.axvline(x=self.R, color='r', linestyle='--', alpha=0.7, label='h = R (metade)')
        ax2.set_xlabel('Altura do Líquido h (m)')
        ax2.set_ylabel('Volume V(h) (m³)')
        ax2.set_title('Volume vs Altura')
        ax2.grid(True, alpha=0.3)
        ax2.legend()
        
        plt.tight_layout()
        plt.show()
        
        return fig, (ax1, ax2)

# Exemplo de uso e teste
if __name__ == "__main__":
    # Criar tanque com R = 1m, L = 5m
    tanque = TanqueCilindricoHorizontal(raio=1.0, comprimento=5.0)
    
    print("=== TANQUE CILÍNDRICO HORIZONTAL ===")
    print(f"Raio (R): {tanque.R} m")
    print(f"Comprimento (L): {tanque.L} m")
    print(f"Volume total: {np.pi * tanque.R**2 * tanque.L:.3f} m³")
    print()
    
    # Testar alguns valores específicos
    alturas_teste = [0, 0.5, 1.0, 1.5, 2.0]
    
    print("TESTES TEÓRICOS - Verificação dos casos:")
    print("h (m) | A(h) (m²) | V(h) (m³) | Estado")
    print("-" * 50)
    
    for h in alturas_teste:
        A = tanque.area_molhada(h)
        V = tanque.volume(h)
        
        if h == 0:
            estado = "Vazio"
        elif h == tanque.R:
            estado = "Metade (teórico: πR²/2)"
        elif h == 2 * tanque.R:
            estado = "Cheio"
        elif h < tanque.R:
            estado = "Inferior"
        else:
            estado = "Superior"
            
        print(f"{h:4.1f} | {A:8.4f} | {V:8.4f} | {estado}")
    
    # Verificação matemática
    print("\nVERIFICAÇÕES MATEMÁTICAS:")
    print(f"A(0) = 0: {tanque.area_molhada(0):.6f}")
    print(f"A(R) = πR²/2 = {np.pi * tanque.R**2 / 2:.6f}: {tanque.area_molhada(tanque.R):.6f}")
    print(f"A(2R) = πR² = {np.pi * tanque.R**2:.6f}: {tanque.area_molhada(2 * tanque.R):.6f}")
    
    # Gerar gráficos
    print("\nGerando gráficos...")
    tanque.plot_curvas()
    
    # Exemplo de cálculo contínuo
    print("\nEXEMPLO PRÁTICO:")
    h_exemplo = 0.75  # 75cm em tanque de 1m raio
    volume_exemplo = tanque.volume(h_exemplo)
    print(f"Para h = {h_exemplo} m: Volume = {volume_exemplo:.3f} m³")
```

__13.Verificação matemática incluida:__

- $A(0) = 0 $ (tanque vazio)

- $A(R) = \pi R^2/2$ ( metade do tanque)

- $A(2R) = \pi R^2$ (tanque cheio)

Como usar:

```python
# Criar tanque
tanque = TanqueCilindricoHorizontal(raio=1.0, comprimento=5.0)

# Calcular volume para altura específica
volume = tanque.volume(0.75)  # h = 0.75m

# Calcular para várias alturas
alturas = [0, 0.5, 1.0, 1.5, 2.0]
volumes = tanque.volume(alturas)

# Gerar gráficos
tanque.plot_curvas()
```
