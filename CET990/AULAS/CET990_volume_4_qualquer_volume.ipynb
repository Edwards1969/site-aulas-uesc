```python
import numpy as np

class TanqueCilindricoHorizontal:
    """
    Classe para cálculo do volume em tanque cilíndrico horizontal
    """
    
    def __init__(self, raio, comprimento):
        self.R = raio
        self.L = comprimento
        
    def area_molhada(self, h):
        """
        Calcula a área molhada da seção transversal para uma altura h
        """
        h = np.asarray(h)
        R = self.R
        
        # Inicializa array de resultados
        A = np.zeros_like(h)
        
        # CASO 1: 0 ≤ h ≤ R (líquido na metade inferior)
        mask_inferior = (h >= 0) & (h <= R)
        if np.any(mask_inferior):
            h_inf = h[mask_inferior]
            termo_arco = R**2 * np.arccos((R - h_inf) / R)
            termo_raiz = (R - h_inf) * np.sqrt(2 * R * h_inf - h_inf**2)
            A[mask_inferior] = termo_arco - termo_raiz
        
        # CASO 2: R < h ≤ 2R (líquido na metade superior)
        mask_superior = (h > R) & (h <= 2 * R)
        if np.any(mask_superior):
            h_sup = h[mask_superior]
            termo_arco_sup = R**2 * np.arccos((h_sup - R) / R)
            termo_raiz_sup = (h_sup - R) * np.sqrt(2 * R * h_sup - h_sup**2)
            area_segmento_superior = termo_arco_sup - termo_raiz_sup
            A[mask_superior] = np.pi * R**2 - area_segmento_superior
        
        # Valores fora do intervalo [0, 2R] retornam NaN
        mask_fora = (h < 0) | (h > 2 * R)
        if np.any(mask_fora):
            A[mask_fora] = np.nan
            
        return A
    
    def volume(self, h):
        """
        Calcula o volume do líquido no tanque para uma altura h
        """
        A = self.area_molhada(h)
        return A * self.L

# EXEMPLO ESPECÍFICO PARA h = 1.2 m
print("=== CÁLCULO PARA ALTURA ESPECÍFICA ===")

# Vamos assumir um tanque com raio de 1.0 m e comprimento de 5.0 m
raio = 1.0      # R = 1.0 m
comprimento = 5.0  # L = 5.0 m
altura_especifica = 1.2  # h = 1.2 m

# Criar o tanque
tanque = TanqueCilindricoHorizontal(raio=raio, comprimento=comprimento)

# Calcular volume para h = 1.2 m
volume = tanque.volume(altura_especifica)

print(f"Parâmetros do tanque:")
print(f"• Raio (R): {raio} m")
print(f"• Comprimento (L): {comprimento} m")
print(f"• Altura do líquido (h): {altura_especifica} m")
print()

# Identificar em qual caso estamos
if altura_especifica <= raio:
    caso = "Caso 1 (0 ≤ h ≤ R) - Líquido na metade inferior"
    formula = "A(h) = R² cos⁻¹((R-h)/R) - (R-h)√(2Rh - h²)"
else:
    caso = "Caso 2 (R < h ≤ 2R) - Líquido na metade superior"
    formula = "A(h) = πR² - [R² cos⁻¹((h-R)/R) - (h-R)√(2Rh - h²)]"

print(f"Situação: {caso}")
print(f"Fórmula aplicada: {formula}")
print()

# Calcular detalhadamente
area_molhada = tanque.area_molhada(altura_especifica)

print(f"RESULTADOS:")
print(f"• Área molhada A(h): {area_molhada:.4f} m²")
print(f"• Volume V(h): {volume:.4f} m³")
print(f"• Volume V(h): {volume:.2f} litros (×1000)")

# Informações adicionais para contexto
volume_total = np.pi * raio**2 * comprimento
percentual_cheio = (volume / volume_total) * 100

print(f"\nCONTEXTO:")
print(f"• Volume total do tanque: {volume_total:.4f} m³")
print(f"• Percentual cheio: {percentual_cheio:.1f}%")

# Verificação da altura relativa
altura_relativa = (altura_especifica / (2 * raio)) * 100
print(f"• Altura em relação ao diâmetro: {altura_relativa:.1f}%")
```
