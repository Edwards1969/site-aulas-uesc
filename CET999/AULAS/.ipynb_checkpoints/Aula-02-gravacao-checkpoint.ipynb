  <strong> Controle PI da Camada de óleo em um Separador Trifásico - Computacional.</strong> 
</div>

Em um Separador Trifásico horizontal, a altura da camada de óleo deve ser mantida próximo do valor de referência $h_{sp}=1.40 \ m$ para garantir a estabilidade da interface gás-óleo e evitar arraste de óleo para a linha de gás. A área transversal do separador é constante e igjal a $A = 8.0 \ m^2$. A vazão de entrada de óleo sofres uma perturbação operacional no instante $t=1 \ h$, conforme a Tabela:

| Tempo (h) | Vazão de entrada de óleo $Q_{o, in}(m^3/h)$ |
|---|-------|
| 0 | 12.0| 
| 1 | 15.0|
| 2 | 15.0|
| 3 | 15.0|
| 4 | 15.0|

A altura da camada de óleo medida pelo sensor é apresentada a seguir:

| Tempo (h) | Altura medida $h_{k}(m)$ |
|---|-------|
| 0 | 1.32  |
| 1 | 1.36  |
| 2 | 1.41  |
| 3 | 1.44  |
| 4 | 1.42  |

A válvula de saída de óleo é controlada por um controlador PI com $K_p=4.0$, $K_1=1.2$, $\Delta t=1 \ h $ , I_o= 0$.

<img src="img/Tanque-separacao-trifasico.png" alt="UESC" style="width: 600px; display: block; margin: 0 auto 5px auto;"><p style="text-align: center;"><strong>Figura: Tanque de Separação de água, óleo e gás proveniente do poço de Petróleo.</strong></p>

Considere as equações:

$ e_k = h_k - h_{sp} $

$ I_{k+1} = I_k  + e_k . \Delta t $

$ Q_{o, out, k} = K_p . e_k + k_i . I_k$

$ \frac{dh}{dt} = \frac{Q_{o, in} -  Q_{o, out}}{A} $

1. Preencha a tabela calculando, para cada instante: o erro $e_k$, a integral acumulada $I_k$ e a vazão de saída calculada pelo controlador $Q_{o, out, k}$;

2. Com base nos valores obtidos, explique se a válvula de saída de óleo tende a abrir ou fechar após a perturbação;

3. Usando a equação dinâmica, calcule a taxa de variação de altura $\frac{dh}{dt}$ nos instante $t=1$ e $t=2$. Explique fisicamente o sinal obtido.

4. Discuta porque esse tipo de cálculo é trabalhoso manualmente e como a simulação computacional (por exemplo, em Python) facilita a análise dinâmica do processo.
