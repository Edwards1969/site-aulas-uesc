Em um Separador Trifásico horizontal, a altura da camada de óleo deve ser mantida próximo do valor de referência $h_{sp}=1.40 \ m$ para garantir a estabilidade da interface gás-óleo e evitar arraste de óleo para a linha de gás. A área transversal do separador é constante e igjal a $A = 8.0 \ m^2$. A vazão de entrada de óleo sofres uma perturbação operacional no instante $t=1 \ h$, conforme a Tabela:

| Tempo (h) | Vazão de entrada de óleo $Q_{o, in}(m^3/h)$ |
|---|-------|
| 0 | 12.0| 
| 1 | 15.0|
| 2 | 15.0|
| 3 | 15.0|
| 4 | 15.0|

A altura da camada de óleo medida pelo sensor é apresentada a seguir:

| Tempo (h) | Altura medida $h_{k}(m)$ |
|---|-------|
| 0 | 1.32  |
| 1 | 1.36  |
| 2 | 1.41  |
| 3 | 1.44  |
| 4 | 1.42  |

A válvula de saída de óleo é controlada por um controlador PI com $K_p=4.0$, $K_1=1.2$, $\Delta t=1 \ h $ , I_o= 0$.

<img src="img/Tanque-separacao-trifasico.png" alt="UESC" style="width: 600px; display: block; margin: 0 auto 5px auto;"><p style="text-align: center;"><strong>Figura: Tanque de Separação de água, óleo e gás proveniente do poço de Petróleo.</strong></p>

Considere as equações:

$ e_k = h_k - h_{sp} $

$ I_{k+1} = I_k  + e_k . \Delta t $

$ Q_{o, out, k} = K_p . e_k + k_i . I_k$

$ \frac{dh}{dt} = \frac{Q_{o, in} -  Q_{o, out}}{A} $

1. Preencha a tabela calculando, para cada instante: o erro $e_k$, a integral acumulada $I_k$ e a vazão de saída calculada pelo controlador $Q_{o, out, k}$;

2. Com base nos valores obtidos, explique se a válvula de saída de óleo tende a abrir ou fechar após a perturbação;

3. Usando a equação dinâmica, calcule a taxa de variação de altura $\frac{dh}{dt}$ nos instante $t=1$ e $t=2$. Explique fisicamente o sinal obtido.

4. Discuta porque esse tipo de cálculo é trabalhoso manualmente e como a simulação computacional (por exemplo, em Python) facilita a análise dinâmica do processo.

__Resolução Computacional:__

```python
# Dados de entrada (extraído do enunciado).
hsp =  1.40   # m
A = 8.0       # m^2
kp = 4.0      # Ganho proporcional
ki = 1.2      # Ganho integral
Delta_t = 1   # horas
I0 = 0
```

__1 - Cálculo do erro:__

$ e_k = h_k - h_{sp} $

```python
# Declarando os valores da vazão de entrada de óleo.
Q_o_in_0, Q_o_in_1, Q_o_in_2, Q_o_in_4, Q_o_in_4 = 12.0, 15.0, 15.0, 15.0, 15.0

# Declarando a altura medida hk (m):
h0, h1, h2, h3, h4 = 1.32, 1.36, 1.41, 1.44, 1.42

e0 = h0 - hsp
e1 = h1 - hsp
e2 = h2 - hsp
e3 = h3 - hsp
e4 = h4 - hsp

# Imprimindo os erros.
print(' O erro  e0 = {:.3f} '.format(e0), 'm')
print(' O erro  e1 = {:.3f} '.format(e1), 'm')
print(' O erro  e2 = {:.3f} '.format(e2), 'm')
print(' O erro  e3 = {:.3f} '.format(e3), 'm')
print(' O erro  e4 = {:.3f} '.format(e4), 'm')
```

__2 - Cálculo da Integral Acumulada.__ 

$ I_{k+1} = I_k + e_k . \Delta t$

```python
# Integrais acumuladas.
I1 = I0 + e0 * Delta_t 
I2 = I1 + e1 * Delta_t
I3 = I2 + e2 * Delta_t  
I4 = I3 + e3 * Delta_t 
I5 = I4 + e4 * Delta_t 

# Imprimindo os resultados.
print('A integral I0 = {:.3f}'.format(I0), 'm.h')
print('A integral I1 = {:.3f}'.format(I1), 'm.h')
print('A integral I2 = {:.3f}'.format(I2), 'm.h')
print('A integral I3 = {:.3f}'.format(I3), 'm.h')
print('A integral I4 = {:.3f}'.format(I4), 'm.h')
```

__3 - Cálculo da ação de controle (Vazão de saída)__

$ Q_{out, k} = k_p . e_k + k_i.I_k$

```python
Q_out_0 = kp * e0 + ki * I0
Q_out_1 = kp * e1 + ki * I1
Q_out_2 = kp * e2 + ki * I2
Q_out_3 = kp * e3 + ki * I3
Q_out_4 = kp * e4 + ki * I4

# Imprimindo os resultados.
print(' O resultado de Q_out_0 = {:.3f}'.format(Q_out_0), 'm^3/h')
print(' O resultado de Q_out_1 = {:.3f}'.format(Q_out_1), 'm^3/h')
print(' O resultado de Q_out_2 = {:.3f}'.format(Q_out_2), 'm^3/h')
print(' O resultado de Q_out_3 = {:.3f}'.format(Q_out_3), 'm^3/h')
print(' O resultado de Q_out_4 = {:.3f}'.format(Q_out_4), 'm^3/h')
```

__4 - Taxa de variação de altura (m/h) dh/dt__

Modelo dinâmico:

$ \frac{dh}{dt} = \frac{Q_{o, in} - Q_{o, out}}{A} $

```python
# Para o tempo de 1h.
dhdt_1h = (Q_o_in_1 - Q_out_1) / A

# Para o tempo de 2h.
dhdt_2h = (Q_o_in_2 - Q_out_2) / A

# Imprimindo os resultados da taxa de variação:
print('A taxa de variação de altura para t=1h é = {:.3f}'.format(dhdt_1h), 'm/h')
print('A taxa de variação de altura para t=2h é = {:.3f}'.format(dhdt_2h), 'm/h')
```

__Interpretação Física:__

Como $\frac{dh}{dt} > 0$ nos dois instantes, a altura da camada de óleo está aumentando. Isso ocorre porque a vazão de entrada ($15.0 \ m^2/h$) é maior que a vazão de saída calculado pelo controlador, que ainda esta reagindo à perturbação ocorrida em $t=1\ h$.
