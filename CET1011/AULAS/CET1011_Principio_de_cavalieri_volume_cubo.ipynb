__Princípio de Cavalieri (intuição)__

O __princípio de Cavalieri__ afirma, em linguagem intuitiva, que dois sólidos têm o mesmo volume se, para toda a altura, suas secções transversais paralelas à base têm áreas iguais. Para volumes por integração, isso equivale a somar (integrar) as áreas das seções transversais ao longo da altura.

__1) Definições e convenções geométricas__

Considere um __cono reto__ de altura ($H$) e raio da base ($R$). Seja um corte por um plano paralelo à base, a uma distância ($x$) do vértice (medida ao longo do eixo). A secção nessa altura é um círculo de raio ($r(x)$). Pela semelhança de triângulos:

<img src="img/tronco_cone_invertido.png" alt="UESC" style="width: 400px; display: block; margin: 0 auto 5px auto;">
<p style="text-align: center;"><strong>Figura: Modelo esquemático de um  tronco de cone invertido.</strong></p>

$$
\frac{r(x)}{x}=\frac{R}{H}

\qquad\Rightarrow\qquad
r(x)=\frac{R}{H}\,x.
$$

A área da secção circular nessa altura é

$$
A(x)=\pi\,r(x)^2=\pi\left(\frac{R}{H}x\right)^2
=\pi\frac{R^2}{H^2}\,x^2.
$$

Pela ideia de Cavalieri, o volume do cone pode ser obtido integrando ($A(x)$) de ($x=0$) até ($x=H$).

__2) Volume do cone por integração (passo a passo)__

$$
V_{\text{cone}}=\int_{0}^{H} A(x)\,dx
=\int_{0}^{H} \pi\frac{R^2}{H^2}x^2\,dx
=\pi\frac{R^2}{H^2}\int_{0}^{H} x^2\,dx.
$$

Calcule a integral:

$$\int_{0}^{H} x^2\,dx=\left.\frac{x^3}{3}\right|_{0}^{H}=\frac{H^3}{3}.$$

Substitua:

$$
V_{\text{cone}}=\pi\frac{R^2}{H^2}\cdot\frac{H^3}{3}
=\frac{1}{3}\pi R^2 H.
$$

Esta é a fórmula clássica do volume de um cone reto.

__3) Volume do tronco de cone (frustum) — dedução por diferença de cones__

Considere agora um __tronco de cone__ (tronco reto), obtido cortando o cone original por um plano paralelo à base. Seja:

- \(H\) = altura do cone maior (do vértice até a base maior),
- \(R\) = raio da base maior,
- Corte paralelo à base a uma distância ($a$) do vértice, gerando uma base interna (menor) de raio ($r_a$),
- A parte entre ($x=a$) e ($x=b$) (com ($0\le a < b \le H$)) forma um tronco com altura ($h=b-a$), base maior de raio ($r_b$) e base menor de raio ($r_a$).

Pelas semelhança:

$$ r_a=\frac{R}{H}\,a,\qquad r_b=\frac{R}{H}\,b. $$

O volume do tronco entre ($x=a$) e ($x=b$) é a diferença entre os volumes do cone até \(b\) e até \(a\):

$$
V_{\text{tronco}}=V_{\text{cone},\,b}-V_{\text{cone},\,a}
=\frac{1}{3}\pi r_b^2 b - \frac{1}{3}\pi r_a^2 a.
$$

Substitua ($r_a$) e ($r_b$):

$$
V_{\text{tronco}}
=\frac{1}{3}\pi\left(\frac{R}{H}b\right)^2 b
-\frac{1}{3}\pi\left(\frac{R}{H}a\right)^2 a
=\frac{\pi R^2}{3H^2}\left(b^3-a^3\right).
$$

Usando fatoração ($b^3-a^3=(b-a)(b^2+ba+a^2)$) e definindo a altura do tronco ($h=b-a$), obtemos a forma clássica:

$$
\boxed{\,V_{\text{tronco}}=\frac{\pi h}{3}\left(r_a^2 + r_a r_b + r_b^2\right)\,}
$$

onde ($r_a=\dfrac{R}{H}a,; r_b=\dfrac{R}{H}b$) e ($h=b-a$). Esta fórmula é equivalente à expressão anterior (pode-se provar substituindo ($r_a,r_b$) e simplificando).

__Observação (Cavalieri)__: o resultado também segue integrando as áreas ($A(x)=\pi\big(\tfrac{R}{H}x\big)^2$) entre ($x=a$) e ($x=b$), que dá precisamente $$(\dfrac{\pi R^2}{3H^2}(b^3-a^3))$$

__4) Volume do cone como caso limite do tronco__

Se colocarmos ($a=0$) e ($b=h$) (tronco que vai do vértice até a altura ($h$), então ($r_a=0$) e a fórmula do tronco reduz-se a

$$V=\frac{\pi h}{3}\left(0^2 + 0\cdot r_b + r_b^2\right)=\frac{\pi h}{3}r_b^2.$$

Como \(r_b=\dfrac{R}{H}h\), obtemos

$$
V=\frac{\pi h}{3}\left(\frac{R}{H}h\right)^2
=\frac{\pi R^2}{3H^2}h^3,
$$

que coincide com a dedução direta do item 2.

__5) Aplicação ao hidrociclone (cone invertido + cilindro) — construção passo a passo__

Adote as convenções do hidrociclone:

- Cone invertido: altura ($H_c$), raio da base ($R$). Origem ($h=0$) no vértice (ponta inferior).
- Cilindro sobreposto: mesma base (raio ($R$)), altura ($H_{\mathrm{cy}}$).
- Nível do fluido: ($h(t)$) medido a partir do vértice do cone para cima.

Para calcular o volume acumulado até \(h\), considere três regimes:

__Regime A — $(0 \le h \le H_c)$ (apenas um cone semelhante preenchido)__

Tronco do vértice até ($h$) é um cone de altura ($h$):

$$V(h)=\frac{\pi R^2}{3H_c^2}\,h^3.$$

__Regime B — $(H_c < h \le H_c+H_{\mathrm{cy}})$ (cone cheio + cilindro parcial)__

Volume do cone cheio:

$$V_c=\frac{1}{3}\pi R^2 H_c.$$

Altura ocupada no cilindro:

$$h_{\mathrm{cy}}=h-H_c.$$

Volume do trecho cilíndrico:

$$V_{\text{cil}}=\pi R^2 (h-H_c).$$

Portanto

$$
V(h)=V_c+\pi R^2(h-H_c)
=\frac{1}{3}\pi R^2 H_c+\pi R^2(h-H_c).
$$

__Regime C__ — $(h>H_c+H_{\mathrm{cy}})$ __(saturado / transbordamento)__

Volume máximo:


$$V_{\max}=V_c+\pi R^2 H_{\mathrm{cy}}.$$

Para $(h>H_c+H_{\mathrm{cy}})$, adote $(V(h)=V_{\max})$.

Em forma por partes:


$$
V(h)=
\begin{cases}
\displaystyle \frac{\pi R^2}{3H_c^2}\,h^3, & 0\le h\le H_c,\\[8pt]
\displaystyle \frac{1}{3}\pi R^2 H_c + \pi R^2(h-H_c), & H_c< h\le H_c+H_{\mathrm{cy}},\\[8pt]
\displaystyle V_{\max}, & h>H_c+H_{\mathrm{cy}}.
\end{cases}
$$


__6) Derivadas temporais — ligação com__ $(\dfrac{dh}{dt})$

Se $(h=h(t))$ é função do tempo, a taxa de variação do volume é:

- Para $(0\le h\le H_c)$:

$$\frac{dV}{dt}=\frac{\pi R^2}{H_c^2}\,h^2\,\frac{dh}{dt}.$$

- Para $(H_c< h\le H_c+H_{\mathrm{cy}})$:


$$\frac{dV}{dt}=\pi R^2\,\frac{dh}{dt}.$$

- Para $(h>H_c+H_{\mathrm{cy}})$:

$$\frac{dV}{dt}=0.$$



__7) Observações pedagógicas (como apresentar no quadro)__

1. Comece apresentando o princípio de Cavalieri: somar áreas das fatias finas (discos) ao longo da altura.  
2. Mostre a semelhança de triângulos para obter $(r(x)=\tfrac{R}{H}x)$.  
3. Faça a integral do disco para deduzir o volume do cone (passo limpo e curto).  
4. Mostre o tronco (frustum) como diferença de dois cones ou integração entre \(a\) e \(b\); deduza a fórmula do tronco e destaque a fatoração $(b^3-a^3=(b-a)(b^2+ba+a^2))$.  
5. Finalmente aplique ao hidrociclone (cone invertido + cilindro), enfatizando os três regimes e escrevendo a expressão por partes para ($V(h)$).  
6. Se desejar, finalize com um pequeno exemplo numérico e um esboço do gráfico \(V\) vs. \(h\) para fixar a intuição.


__Dedução didática e passo a passo (cone invertido + cilindro)__ 

Introdução e objetivo
Neste roteiro didático vamos deduzir, **passo a passo**, a expressão do **volume de fluido** dentro de um hidrociclone cuja geometria é a união de:

- um **cone reto invertido** (vértice para baixo), de altura \(H_c\) e raio da base \(R\); e  
- um **cilindro** apoiado sobre a base do cone, com o mesmo raio \(R\) e altura \(H_{\mathrm{cy}}\).

Assumimos que a **origem da coordenada vertical** está no vértice do cone (ponta inferior), de modo que a altura do nível de fluido é dada por \(h(t)\) medida para cima a partir da ponta. O problema: **determinar o volume de fluido \(V(h)\)** abaixo da linha de nível \(h\), de forma que um professor possa expô-lo no quadro em blocos lógicos bem definidos.

Bloco 1 — Princípio de Cavalieri (intuição essencial)
1. Enuncie: *"Se dois sólidos têm, para toda altura, áreas de secção transversal iguais, então têm igual volume."*  
2. Interprete para nosso caso: o volume do sólido abaixo de uma altura \(h\) pode ser obtido somando (integrando) as áreas das secções transversais circulares ao longo do eixo vertical desde \(0\) até \(h\).  
3. Pedagógico no quadro: desenhe um pequeno cilindro elemental de espessura \(dz\) e área \(A(z)\); mostre que o volume elemental é \(dV = A(z)\,dz\) e, portanto, \(V=\int A(z)\,dz\).


__Bloco 2 — Semelhança de triângulos: raio da secção em função da altura (cone)__

1. Desenhe o cone invertido e marque: vértice (origem), altura total do cone \(H_c\), raio da base \(R\).  
2. Considere uma secção a uma altura \(x\) (do vértice). Pelo desenho e semelhança de triângulos:
   
$$
\frac{r(x)}{x}=\frac{R}{H_c}
\quad\Rightarrow\quad
r(x)=\frac{R}{H_c}\,x.
$$
  
3. Interprete: quando ($x=0$) então ($r=0$); quando ($x=H_c$) então ($r=R$). Mostre no quadro os triângulos semelhantes.

__Bloco 3 — Área da secção e integral (cone)__

1. A área da secção circular a ($x$) é

$$
A(x)=\pi [r(x)]^2
   =\pi\left(\frac{R}{H_c}x\right)^2
   =\pi\frac{R^2}{H_c^2}\,x^2.
$$
   
__2. Pelo princípio de Cavalieri, o volume do cone entre \(x=0\) e \(x=h\) é__


$$
V(h)=\int_{0}^{h} A(x)\,dx
   =\int_{0}^{h} \pi\frac{R^2}{H_c^2}\,x^2\,dx
   =\pi\frac{R^2}{H_c^2}\int_{0}^{h} x^2\,dx.
$$

__3. Calcule a integral (faça passo a passo no quadro):__

$$
\int_{0}^{h} x^2\,dx=\left.\frac{x^3}{3}\right|_0^h=\frac{h^3}{3}.
$$

__4. Conclua a expressão para o volume do cone parcial__ (até altura \(h\), válido enquanto ($0\le h\le H_c$)):

$$
\boxed{\,V(h)=\frac{\pi R^2}{3H_c^2}\,h^3\qquad(0\le h\le H_c).\,}
$$

5. __Observação didática: explique que esse__ ($V(h)$) __é o volume de um cone semelhante (mesma forma, escala menor) de altura __\(h\).


__Bloco 4 — Volume do tronco (frustum) — ligação com o cone__

1. Explique: o volume do trecho do cone entre duas alturas \(a\) e \(b\) $$((0\le a<b\le H_c))$$ é a diferença entre dois cones:

$$
V_{\text{tronco}}=\frac{\pi R^2}{3H_c^2}\,(b^3-a^3).
$$

2. Mostre a fatoração $(b^3-a^3=(b-a)(b^2+ba+a^2))$ e a forma clássica do tronco:

$$
V_{\text{tronco}}=\frac{\pi h}{3}\left(r_a^2 + r_a r_b + r_b^2\right),
$$

onde ($h=b-a$), ($r_a=\dfrac{R}{H_c}a$) e $(r_b=\dfrac{R}{H_c}b)$.

3. Pedagógico: contraste as duas formas e proponha um exercício rápido: obter o volume entre \(a\) e \(b\) por integração para confirmar a fórmula.

__Bloco 5 — Volume do cilindro (fatia adicional)__

1. Recorde a fórmula do volume de um cilindro de raio \(R\) e altura \(h\):

$$
V_{\text{cil}}(h)=\pi R^2 h.
$$   

2. No contexto do hidrociclone, quando o fluido ultrapassa a altura do cone \(H_c\), a porção adicional dentro do cilindro (altura \(h-H_c\)) tem volume:

$$
V_{\text{cil, extra}}=\pi R^2 (h-H_c).
$$

3. Pedagógico: desenhar a linha de nível no cone + cilindro e destacar visualmente o trecho cilíndrico.

__Bloco 6 — União das partes: expressão por regimes (volume total até altura \(h\))__

1. **Regime A — apenas cone preenchido** $((0\le h\le H_c))$:

$$
V(h)=\frac{\pi R^2}{3H_c^2}\,h^3.
$$

2. __Regime B — cone cheio + cilindro parcialmente preenchido__ $((H_c< h\le H_c+H_{\mathrm{cy}}))$:

$$
V(h)=V_c+\pi R^2(h-H_c),
\quad\text{onde}\quad V_c=\frac{1}{3}\pi R^2 H_c.
$$

Ou seja,

$$
\boxed{\,V(h)=\frac{1}{3}\pi R^2 H_c + \pi R^2 (h-H_c)\quad (H_c< h\le H_c+H_{\mathrm{cy}}).\,}
$$

3. **Regime C — transbordamento / saturado** $((h>H_c+H_{\mathrm{cy}}))$:

$$
V(h)=V_{\max}=V_c+\pi R^2 H_{\mathrm{cy}}.
$$

4. Reescreva compactamente em forma por partes:

$$
V(h)=
\begin{cases}
\displaystyle \frac{\pi R^2}{3H_c^2}\,h^3, & 0\le h\le H_c,\\[8pt]
\displaystyle \frac{1}{3}\pi R^2 H_c + \pi R^2 (h-H_c), & H_c< h\le H_c+H_{\mathrm{cy}},\\[8pt]
\displaystyle V_{\max}, & h>H_c+H_{\mathrm{cy}}.
\end{cases}
$$

 Bloco 7 — Derivadas temporais (ligando com dh/dt) — interpretação física

1. Se o nível \(h\) varia no tempo: \(h=h(t)\), então a taxa de variação do volume \(dV/dt\) é obtida por derivação das expressões acima:

- Para $(0\le h\le H_c)$:

$$
\frac{dV}{dt}=\frac{\pi R^2}{H_c^2}\,h^2\,\frac{dh}{dt}.
$$

- Para $(H_c< h\le H_c+H_{\mathrm{cy}})$:

$$
\frac{dV}{dt}=\pi R^2\,\frac{dh}{dt}.
$$

- Para $(h>H_c+H_{\mathrm{cy}})$:

$$
\frac{dV}{dt}=0.
$$

2. Interprete no quadro: quando o nível está no cone, um pequeno aumento \(dh\) acrescenta volume proporcional a $(h^2\,dh)$; no cilindro, o acréscimo é constante $(\pi R^2 dh)$.

__Bloco 8 — Inversa__ $(h(V))$ __(recuperar nível a partir do volume) — útil para medições__

1. Se quiser obter \(h\) a partir de um volume \(V\):

- Se $(0\le V\le V_c)$ (só cone parcial):

$$
V=\frac{\pi R^2}{3H_c^2}h^3
\quad\Rightarrow\quad
h(V)=\left(\frac{3H_c^2}{\pi R^2}V\right)^{1/3}.
$$

- Se \(V_c < V \le V_{\max}\) (cone cheio + cilindro parcial):

$$
V=V_c+\pi R^2(h-H_c)
\quad\Rightarrow\quad
h(V)=H_c+\frac{V-V_c}{\pi R^2}.
$$

2. Pedagógico: comente que a inversa é útil quando se mede volume/possível massa e precisa-se saber o nível \(h\).

__Bloco 9 — Sugestão de sequência didática para o quadro (tempo estimado por bloco)__

1. Desenho + enunciado do problema (2–3 min).  
2. Princípio de Cavalieri e disco elemental (3–4 min).  
3. Semelhança de triângulos e $(r(x))$ (3 min).  
4. Integração e dedução de $(V(h))$ para o cone (5–7 min).  
5. Tronco de cone como diferença — comentário e exercício (4–5 min).  
6. Introdução do cilindro e soma das partes (4–5 min).  
7. Derivadas temporais e interpretação (3–4 min).  
8. Exemplo numérico rápido no quadro (5–8 min).  

(Adapte os tempos conforme a dinâmica da turma; os minutos são apenas indicativos.)

__Bloco 10 — Checkpoints de compreensão (perguntas rápidas para os alunos)__

1. Por que usamos semelhança de triângulos para encontrar $(r(x))$?  
2. O que significa integrar as áreas $(A(x))$? (ligação com o Princípio de Cavalieri)  
3. Por que a fórmula muda quando $(h>H_c)$?  
4. Como muda $(dV/dt)$ quando o nível passa do cone para o cilindro?  
5. Peça para deduzirem $(V(h))$ para um conjunto numérico simples (ex.: $(R=0{,}2) m$, $(H_c=0{,}8) m$).

---

__Observações finais para o professor__

- Mantenha a explicação em **blocos**: semelhança geométrica → área da secção → integração → soma com o cilindro.  
- Use sempre um **desenho grande e claro**: marque $(H_c)$, $(R)$, $(H_{\mathrm{cy}})$, e uma linha para $(h)$.  
- Se a turma tiver mais facilidade com álgebra do que com integração, apresente a **fórmula conhecida do volume do cone** como um atalho (e mostre a integração como justificativa).  
- Para turmas com menos formação em cálculo, foque na intuição pelo tronco (diferença de volumes) e no conceito de secções.
