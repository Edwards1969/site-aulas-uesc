__Parâmetros numéricos adotados (Projeto)__

$$
H_c = 1{,}20\ \text{m}, \qquad
H_{\mathrm{cy}} = 0{,}80\ \text{m}, \qquad
R = 0{,}35\ \text{m}.
$$
	
__1. Relações geométricas e dedução simbólica__

Raio da seção em função da altura (semelhança de triângulos).

$$
r(h)=\frac{R}{H_c}\,h.
$$

Área da seção circular.

$$
A(h)=\pi r(h)^2 = \pi \left(\frac{R}{H_c}h\right)^2 = \pi\frac{R^2}{H_c^2}\,h^2.
$$
	
Volume do cone preenchido até altura \(h\) (princípio de Cavalieri / integração)

Para $(0\le h\le H_c)$,

$$
V(h)=\int_0^h A(x)\,dx
=\int_0^h \pi\frac{R^2}{H_c^2}x^2\,dx
=\pi\frac{R^2}{H_c^2}\cdot\frac{h^3}{3}
=\frac{\pi R^2}{3H_c^2}\,h^3.
$$

Volume do cone total (quando $(h=H_c)$)

$$
V_c = \frac{1}{3}\pi R^2 H_c.
$$

Para $(h>H_c)$ (cone cheio + trecho do cilindro).	Altura ocupada no cilindro:

$$
h_{\mathrm{cy}}=h-H_c.
$$

Volume adicional no cilindro:

$$ 
V_{\text{cil,extra}}=\pi R^2 (h-H_c). 
$$

Volume total (para $(H_c< h \le H_c+H_{\mathrm{cy}})$):

$$ 
V(h)=V_c+\pi R^2 (h-H_c).
$$

Volume máximo (saturado / transbordamento).

$$
V_{\max}=V_c+\pi R^2 H_{\mathrm{cy}}.
$$

Expressão final por partes

$$
V(h)=
\begin{cases}
	\dfrac{\pi R^2}{3H_c^2}\,h^3, & 0\le h\le H_c,\\[8pt]
	\dfrac{1}{3}\pi R^2 H_c + \pi R^2 (h-H_c), & H_c< h\le H_+H_{\mathrm{cy}},\\[8pt]
	V_{\max}, & h>H_c+H_{\mathrm{cy}}.
\end{cases}
$$


__2. Derivadas temporais (liga com $(\dot h = dh/dt)$)__

$$
\frac{dV}{dt} =
\begin{cases}
		\displaystyle \frac{\pi R^2}{H_c^2}\,h^2\,\frac{dh}{dt}, & 0\le h\le H_c,\\[8pt]
		\displaystyle \pi R^2\,\frac{dh}{dt}, & H_c< h\le H_c+H_{\mathrm{cy}},\\[8pt]
		0, & h>H_c+H_{\mathrm{cy}}.
\end{cases}
$$

__3. Inversa $(h(V))$ (recuperar nível a partir do volume)__

Relação entre $(V)$ e $(h(V))$.

- Se $(0 \le V \le V_c)$:

$$
V = \frac{\pi R^2}{3H_c^2} \, h^3
$$

$$
h(V) = \left( \frac{3H_c^2}{\pi R^2} \, V \right)^{1/3}
$$

- Se $(V_c < V \le V_{\max})$:

$$
V = V_c + \pi R^2 (h - H_c)
$$

$$
h(V) = H_c + \frac{V - V_c}{\pi R^2}
$$


__4. Cálculos numéricos detalhados (passo a passo)__

Calcularemos as constantes auxiliares com aritmética mostrada por etapas.

1) $(R^2)$.
	
$$
R^2 = (0{,}35)^2 = 0{,}35\times 0{,}35 = 0{,}1225.
$$
	
	
2) $(\pi R^2)$ (área da base circular).
	
$$
\pi R^2 = \pi \times 0{,}1225 \approx 3{,}141592653589793\times 0{,}1225
=0{,}3848451000647496\ \text{m}^2.$$
	
3) $(H_c^2)$.
	
$$
H_c^2 = (1{,}20)^2 = 1{,}44.
$$
	
4) Coeficiente do termo cúbico do cone:

$$
\frac{\pi R^2}{3H_c^2}
= \frac{0{,}3848451000647496}{3\times 1{,}44}
= \frac{0{,}3848451000647496}{4{,}32}
\approx 0{,}08908451390387721.
$$

Chamaremos este valor de $(C_{\text{cone}})$.
	
$$
C_{\text{cone}} \approx 0,08908451390387721\ m^{-3}.
$$.

5) Volume do cone total $(V_c)$.

Usando $(V_c=\dfrac{1}{3}\pi R^2 H_c)$:

$$

	V_c = \frac{1}{3}\times 0{,}3848451000647496 \times 1{,}20
	= 0{,}3848451000647496\times 0{,}4
	= 0{,}15393804002589984\ \text{m}^3.
	
$$
		
6) Volume máximo $(V_{\max})$ (cone + cilindro cheio).

$$
V_{\max}=V_c+\pi R^2 H_{\mathrm{cy}}
=0{,}15393804002589984 + 0{,}3848451000647496\times 0{,}80.
$$
	
Calcule a parte cilíndrica:

$$
0{,}3848451000647496\times 0{,}80 = 0{,}3078760800517997.
$$

Somando:

$$V_{\max}=0{,}15393804002589984 + 0{,}3078760800517997
=0{,}46181412007769954 \approx 0{,}4618141200776995\ \text{m}^3.$$
	

__5. Exemplos numéricos (valores de prova) — cálculo passo a passo__

h_2^3=(0{,}80)^3=0{,}512.

$$
V(0{,}80)=C_{\text{cone}}\times 0{,}512
	=0{,}08908451390387721\times 0{,}512

Calcule a altura no cilindro:
	
$$

$$
	
Volume cilíndrico adicional:

$$

Somando com $(V_c)$:
	
$$
V(1{,}50)=0{,}15393804002589984 + 0{,}11545353001942488
	=0{,}26939157004532472\ \text{m}^3.
	$$

$$
\boxed{V(1{,}50)\approx 0{,}26939157\ \text{m}^3.}

```python
import numpy as np

# ----------------------------------------------------------
# PARÂMETROS DO HIDROCICLONE
# ----------------------------------------------------------
R_top = 0.30      # raio do cone na parte superior (m)
R_bottom = 0.05   # raio do cone na ponta (m)
H_cone = 1.00     # altura total do cone (m)
H_cyl  = 1.50     # altura do cilindro (m)

# ----------------------------------------------------------
# FUNÇÕES GEOMÉTRICAS
# ----------------------------------------------------------

def raio_no_cone(h):
    """
    Calcula o raio do nível h dentro do cone.
    Relação linear pelo triângulo semelhante.
    """
    return R_bottom + (R_top - R_bottom) * (h / H_cone)

def volume_no_cone(h):
    """
    Volume acumulado dentro do cone até a altura h.
    Fórmula do tronco de cone.
    """
    r = raio_no_cone(h)
    return (np.pi * h * (R_bottom**2 + r**2 + R_bottom * r)) / 3

def volume_total_cone():
    """Volume do cone completo."""
    return volume_no_cone(H_cone)

def volume_no_cilindro(h):
    """
    Volume acumulado no cilindro a partir do topo do cone.
    h é altura dentro do cilindro, não altura total.
    """
    return np.pi * (R_top**2) * h

def volume_total_cilindro():
    """Volume do cilindro completo."""
    return volume_no_cilindro(H_cyl)

def volume_total_hidrociclone():
    """Volume máximo: cone cheio + cilindro cheio."""
    return volume_total_cone() + volume_total_cilindro()
```

```python
# ----------------------------------------------------------
# FUNÇÃO PRINCIPAL: volume dado h(t)
# ----------------------------------------------------------

def volume(h):
    """
    Calcula o volume até uma altura TOTAL h:
    - Se h ≤ H_cone  → nível está no cone
    - Se h > H_cone  → soma cone cheio + parte do cilindro
    """
    if h <= H_cone:
        return volume_no_cone(h)
    else:
        return volume_total_cone() + volume_no_cilindro(h - H_cone)

# ----------------------------------------------------------
# TESTE: altura desejada pelo usuário
# ----------------------------------------------------------

h_values = [0.40, 0.80, 1.50]     # alturas que você informou

for h in h_values:
    print(f"Volume para h = {h:.2f} m → {volume(h):.6f} m³")

# ----------------------------------------------------------
# RESULTADOS GERAIS
# ----------------------------------------------------------
print("\n==================== RESULTADOS GERAIS ====================")
print(f"Volume total do cone .............: {volume_total_cone():.6f} m³")
print(f"Volume total do cilindro .........: {volume_total_cilindro():.6f} m³")
print(f"Volume máximo do hidrociclone ....: {volume_total_hidrociclone():.6f} m³")
print("===========================================================")
```
