__Hidrociclones na Indústria de Petróleo__

__O que é um Hidrociclone?__

Um __hidrociclone__ é um equipamento de separação que utiliza a __força centrífuga__ para promover a separação de partículas ou gotas presentes em um fluido. Trata-se de um dispositivo sem partes móveis, o que lhe confere elevada robustez e confiabilidade operacional. Sua geometria clássica é composta por duas regiões principais:

- __Parte superior cilíndrica__, onde ocorre a entrada tangencial do fluido.
- __Parte inferior cônica__, cuja área diminui progressivamente até o vértice.
\end{itemize}

A combinação dessas duas regiões forma um escoamento em vórtice capaz de gerar forças centrífugas elevadas, tornando o processo de separação rápido e eficiente.

Um  __hidrociclone__ é um equipamento de separação que utiliza a __força centrífuga__ para promover a separação de partículas ou gotas presentes em um fluido. Trata-se de um dispositivo sem partes móveis, o que lhe confere elevada robustez e confiabilidade operacional. Sua geometria clássica é composta por duas regiões principais:

- __Parte superior cilíndrica__, onde ocorre a entrada tangencial do fluido.
- __Parte inferior cônica__, cuja área diminui progressivamente até o vértice.


A combinação dessas duas regiões forma um escoamento em vórtice capaz de gerar forças centrífugas elevadas, tornando o processo de separação rápido e eficiente.

__Como o Hidrociclone Funciona__

Quando o fluido entra tangencialmente na parte cilíndrica, desenvolve-se um movimento rotacional de alta velocidade, formando um __vórtice helicoidal__. O escoamento interno é geralmente descrito como composto por dois vórtices:

- __Vórtice externo descendente__: transporta partículas ou gotas maiores e mais densas para a parede interna do equipamento. Essas partículas são conduzidas para a parte inferior, onde deixam o hidrociclone pelo __underflow__.
- __Vórtice interno ascendente__: forma-se próximo ao eixo central. Partículas ou gotas menores e menos densas tendem a migrar para essa região e ascendem até a saída superior, denominada __overflow__.
\end{enumerate}

Esse mecanismo ocorre porque a força centrífuga atua com maior intensidade sobre partículas de maior diâmetro ou maior densidade, empurrando-as radialmente para as regiões externas do escoamento. Partículas menores, ou com menor densidade, permanecem mais próximas do eixo central, sendo então conduzidas para o __overflow__.

__Aplicações do Hidrociclone__

O hidrociclone é amplamente utilizado para promover separações envolvendo diferenças de densidade ou de tamanho de partículas, tais como:


-sólidos em suspensão;
- gotas de água em óleo;
- gotas de óleo em água;
- misturas contendo partículas com diferentes granulometrias.


__Importância para a Indústria de Petróleo__

Durante o processo de produção, o petróleo extraído de um reservatório carrega consigo partículas de água, gotas de óleo, sólidos finos e outras impurezas. A remoção rápida e contínua desses componentes é fundamental para assegurar:

- proteção de equipamentos subsequentes, como bombas, válvulas e trocadores de calor;
- melhoria na qualidade da água produzida antes do descarte ou reinjeção;
  redução da formação de emulsões estáveis;
- diminuição da necessidade de grandes tanques de separação gravitacional;
- maior compacidade dos sistemas de tratamento, especialmente em plataformas offshore.

O hidrociclone atende a essas necessidades devido às elevadas forças centrífugas geradas no interior do equipamento, que podem ser centenas ou milhares de vezes superiores à gravidade. Esse mecanismo possibilita que:

- __partículas maiores__ ou mais densas sejam direcionadas para a parede do equipamento e removidas pelo __underflow__;
- __partículas menores ou menos densas__ sejam transportadas para o eixo do vórtice e removidas pelo __overflow__.

Esse comportamento natural do escoamento faz do hidrociclone um dos dispositivos mais eficientes, compactos e confiáveis para o pré-tratamento de petróleo, água produzida e efluentes da indústria de óleo e gás.

__Aplicações típicas no petróleo e gás__

__1. Tratamento de ágaua oleosa produzida__

- Remoção de g´tículas de óleo da água produzida
- Pode trabalhar em série com flotação por gás e separadores gravitacionais.

__2. Separação de areia de produção__
- Área crítica em poços subtmarinos e FPSOs.
- Concentração de descarte de sólidos arrastados do reservatório.

__3. Pré-tratamento em processos de separação trifásico__
- Complemento a sepaadores horizontais e vasos de produção.

__4. Emulsões e efluentes complexos.__
- Alguns modelos especializados são sados para quebrar emulsóes água-óleo com diferentes faixas granulométricas.


__Como funciona fisicamente__

O princípio é baeado na criação de um movimento helicoidal:
- Velcidade tangencial -> geração de forçca cetrífuga.
- O fluido forma um vórtice duplo:
    - um externo (descendente)
    - um interno (ascendente)
    
-A separação ocorre pela diferença entre:

$$ F_c = m \frac{v^2}{r} $$

onde partículas mais pesadas (ou gotas maiores)sofrem maior força centrifuga e migram para a parede.

__Vantagens__
- Sem partes móveis -> baixa manutenção.
- Compacto, leve, bom para FPSOs
- Alta eficiência para partículas pequenas (20-40 $\mu m$).

__Dedução matemática__

Imagine um hidrociclone como mostrado na figura abaixo. Ele é composto por duas regiões geométricas:

1. Um **cone invertido**, com altura \(H_c\) e raio da base \(R\);
2. Um **cilindro** com o mesmo raio \(R\) e altura \(H_{cy}\), apoiado sobre o cone.

Adotamos um sistema de coordenadas em que a origem está no vértice do cone, ou seja, \(H = 0\) na ponta inferior. Assim, ao subirmos pelo equipamento, a altura total do fluido é representada por \(h(t)\).

Nosso objetivo é determinar **o volume de fluido abaixo da altura \(h\)**, expresso como função de \(h\).

A figura abaixo ilustra o hidrociclone idealizado, com suas principais dimensões.

2) Parte 1 — Geometria do cone (intuição)

Antes de escrever fórmulas, observe que no cone o raio não é constante.  
Quanto maior a altura \(h\), maior será o raio da seção circular naquele ponto.

Como o cone é uma figura semelhante a si mesma em qualquer altura, temos a relação de semelhança:

$$\frac{r(h)}{h} = \frac{R}{H_c}$$

Observe que quando $h = H_c$ temos $r=R$; quando $h=0$ temos $r=0.
$
Portanto, o raio na altura \(h\) é:

$$r(h) = \frac{R}{H_c} \, h$$

A área da seção transversal a essa altura é circular:

$$
A(h) = \pi \, r(h)^2 = \pi \left( \frac{R}{H_c} h \right)^2
$$

Logo:

$$A(h) = \pi \frac{R^2}{H_c^2} h^2$$

E o volume acumulado até a altura \(h\) é a integral das áreas:

$$
V(h) = \int_{0}^{h} A(\xi)\, d\xi
$$


Substituindo:


$$
V(h) = \int_0^h \pi \frac{R^2}{H_c^2}\xi^2 \, d\xi
$$



$$
V(h) = \pi \frac{R^2}{H_c^2} \left[ \frac{\xi^3}{3} \right]_{0}^{h}
$$



$$
\boxed{V(h) = \frac{\pi R^2}{3H_c^2} h^3}
$$


Essa expressão vale apenas para:


$$
0 \le h \le H_c
$$


Pois acima disso já entramos no cilindro.

__3) Parte 2 — Quando o fluido ultrapassa o cone e entra no cilindro__

Se a altura do fluido é maior que a altura total do cone \((h > H_c)\), então temos duas contribuições:

1. O volume total do cone (completamente cheio):
   
   $$
   V_{\text{cone}} = \frac{1}{3} \pi R^2 H_c
   $$
   

2. O volume cilíndrico adicional:
   
   $$
   V_{\text{cilindro}} = \pi R^2 (h - H_c)
   $$
   

Portanto, para \(h \ge H_c\):


$$
\boxed{
V(h) = \frac{1}{3}\pi R^2 H_c + \pi R^2 (h - H_c)
}
$$


Pode-se reescrever de forma mais elegante:


$$
V(h) = \pi R^2 \left( h - \frac{2}{3}H_c \right)
$$


__4) Expressão final por partes__


$$V(h) =
\begin{cases}
\displaystyle \frac{\pi R^2}{3H_c^2} h^3, & 0 \le h \le H_c, \\[12pt]
\displaystyle \frac{1}{3}\pi R^2 H_c + \pi R^2 (h - H_c), & H_c < h \le H_c + H_{cy}.
\end{cases}$$


Essa função por partes permite calcular o volume de fluido no hidrociclone para qualquer altura \(h(t)\).

__Vamos para uma dedução mais completa — passo a passo__

**Convenções:** considere um hidrociclone composto por um **cone invertido** (vértice para baixo) de altura \(H_c\) e raio de base \(R\), sobreposto por um **cilindro** de mesmo raio \(R\) e altura \(H_{\mathrm{cy}}\). A origem da coordenada vertical é tomada no vértice do cone, ou seja, \(h=0\) na ponta inferior. O nível de fluido é \(h(t)\), medido a partir da ponta do cone para cima.


__1. Volume como função de \(h\) — derivação__

__(a) Parte cônica — $(0 \le h \le H_c)$

Por semelhança de triângulos, o raio à altura \(h\) (medida da ponta) escala linearmente:


$$r(h)=\frac{R}{H_c}\,h.$$


O volume até a altura \(h\) é a soma (integral) de discos de área \(\pi r(z)^2\):


$$
V(h)=\int_{0}^{h} \pi\,[r(z)]^2\,dz
= \int_{0}^{h} \pi \left(\frac{R}{H_c} z\right)^2 dz
= \pi\frac{R^2}{H_c^2}\int_{0}^{h} z^2\,dz.
$$

Calculando a integral:

$$\int_{0}^{h} z^2\,dz = \frac{h^3}{3},$$

portanto

$$
\boxed{\,V(h)=\frac{\pi R^2}{3H_c^2}\;h^3\qquad(0\le h\le H_c)\,}.
$$

(Observação: alternativamente, usar a fórmula conhecida do volume de um cone $(\tfrac{1}{3}\pi r^2 h)$ com $(r=\tfrac{R}{H_c}h)$ leva à mesma expressão.)


__(b) Cone cheio + cilindro parcial__ — $(H_c < h \le H_c+H_{\mathrm{cy}})$

O volume total do cone é

$$
V_c=\frac{1}{3}\pi R^2 H_c.
$$


Se ($h>H_c$), a porção adicional no cilindro tem altura ($h-H_c$) e volume ($\pi R^2(h-H_c)$). Assim,


$$\boxed{\,V(h)=V_c+\pi R^2(h-H_c)
=\frac{1}{3}\pi R^2 H_c+\pi R^2(h-H_c)\,}
\qquad\text{para }H_c< h\le H_c+H_{\mathrm{cy}}.$$


__c) Transbordamento — ($h>H_c+H_{\mathrm{cy}}$)__

O volume máximo (tanque cheio) é


$$
V_{\max}=V_c+\pi R^2 H_{\mathrm{cy}}.
$$


Para \(h>H_c+H_{\mathrm{cy}}\) considera-se o volume saturado:


$$
\boxed{\,V(h)=V_{\max}\,.}
$$


__2. Forma única (compacta) usando recortes__

Defina

$$
h_c=\min(h,H_c),\qquad h_{\mathrm{cy}}=\max\big(0,\min(h-H_c,H_{\mathrm{cy}})\big).
$$

Então a expressão compacta conveniente para implementação numérica é


$$
\boxed{\,V(h)=\frac{\pi R^2}{3H_c^2}\,h_c^3 \;+\; \pi R^2\,h_{\mathrm{cy}}\,.}
$$


__3. Taxa de variação temporal \(\dfrac{dV}{dt}\) (em termos de \(\dfrac{dh}{dt}\))__

Derivando peça a peça:

- Para ($0\le h\le H_c$):


$$
\frac{dV}{dt}
= \frac{d}{dt}\!\left(\frac{\pi R^2}{3H_c^2}h^3\right)
= \frac{\pi R^2}{H_c^2}\;h^2\;\frac{dh}{dt}.
$$


- Para ($H_c< h\le H_c+H_{\mathrm{cy}}$):


$$
\frac{dV}{dt}
= \frac{d}{dt}\!\left(V_c+\pi R^2(h-H_c)\right)
= \pi R^2\;\frac{dh}{dt}.
$$

- Para ($h>H_c+H_{\mathrm{cy}}$):

$$\frac{dV}{dt}=0.$$

Forma compacta com funções indicadoras

$$
\frac{dV}{dt}
=\frac{\pi R^2}{H_c^2}\,h^2\frac{dh}{dt}\,\mathbf{1}_{\{0\le h\le H_c\}}
+\pi R^2\frac{dh}{dt}\,\mathbf{1}_{\{H_c< h\le H_c+H_{\mathrm{cy}}\}}.
$$

4. Inversa ($h(V)$) — recuperar a altura a partir do volume

- Se ($0\le V\le V_c$) (apenas cone parcial):

$$
V=\frac{\pi R^2}{3H_c^2}h^3
\quad\Rightarrow\quad
\boxed{\,h(V)=\left(\frac{3H_c^2}{\pi R^2}\,V\right)^{1/3}.}
$$

- Se ($V_c < V \le V_{\max}$) (cone cheio + cilindro parcial):


$$
V=V_c+\pi R^2 (h-H_c)
\quad\Rightarrow\quad
\boxed{\,h(V)=H_c+\frac{V-V_c}{\pi R^2}.}
$$


- Se ($V>V_{\max}$): considera-se ($h=H_c+H_{\mathrm{cy}}$) (saturado / transbordamento).

__5. Observações práticas para a prova / implementação__

1. **Referência de \(h\):** confirme se \(h=0\) corresponde ao vértice do cone (adotado aqui). Caso a referência seja diferente, ajuste as fórmulas aplicando a mudança de origem apropriada.  
2. **Implementação numérica:** utilize a forma com \(\min\)/\(\max\) (ou `clip`) para evitar ramificações e facilitar vetorização.  
3. **Condições iniciais:** se for dada \(h(t)\) ou ($\dfrac{dh}{dt}$), use as fórmulas acima para obter \(V(t)\) e ($\dfrac{dV}{dt}$).  
4. **Unidades:** mantenha coerência ($m$, $m^2$, $m^3$, $s$).  
5. **Nota de modelagem:** as fórmulas tratam apenas do volume geométrico preenchido; para modelar fenômenos como concentração da “nuvem de fluido” ou mistura, será necessário acoplar balanços de massa/energia e resolver equações adicionais.

__Exemplo numérico__

Vamos considerar um hidrociclone com as seguintes dimensões:


$$
R = 0.20 \text{ m}, \qquad
H_c = 0.80 \text{ m}, \qquad
H_{\mathrm{cy}} = 1.20 \text{ m}.
$$


O objetivo é calcular o **volume de fluido** para três valores de altura:

1. ($h_1 = 0.40\text{ m}$) — dentro do cone  
2. ($h_2 = 0.80\text{ m}$) — topo do cone  
3. ($h_3 = 1.50\text{ m}$) — dentro do cilindro


__1) Caso 1 — ponto dentro do cone ($(h_1 = 0.40\text{ m})$)__

Como ($h_1 < H_c$), usamos:


$$V(h)=\frac{\pi R^2}{3H_c^2}h^3.$$


Substituindo:


$$V(0.40)=\frac{\pi (0.20)^2}{3(0.80)^2}(0.40)^3.$$


Cálculo passo a passo:


$$
0.20^2 = 0.04,
\qquad
0.80^2 = 0.64,
\qquad
0.40^3 = 0.064.
$$



$$V(0.40)=\frac{\pi \cdot 0.04}{3\cdot0.64}\cdot0.064.$$



$$V(0.40)=\frac{0.04\pi}{1.92}\cdot0.064.$$



$$V(0.40)=0.02094 \cdot 0.064\pi.$$



$$\boxed{V(0.40)\approx 0.00421 \text{ m}^3.}$$


__2) Caso 2 — topo do cone ($(h_2 = 0.80\text{ m})$)__

O volume do cone cheio é:

$$V_c = \frac{1}{3}\pi R^2 H_c.$$

Substituindo:

$$
V_c=\frac{1}{3}\pi (0.20)^2(0.80)
     =\frac{1}{3}\pi(0.04)(0.80)
     =\frac{1}{3}\pi(0.032).
$$

$$\boxed{V_c \approx 0.03351 \text{ m}^3.}$$

Como ($h_2 = H_c$), este é o volume buscado.

__3) Caso 3 — fluido dentro do cilindro ($(h_3 = 1.50\text{ m})$)__

Primeiro verifique:

$$
h_3 > H_c \quad\Rightarrow\quad \text{há cone cheio + cilindro parcial}.
$$

Altura dentro do cilindro:

$$h_{\mathrm{cy}} = h_3 - H_c = 1.50 - 0.80 = 0.70\text{ m}.$$

Volume total:

$$ V(h_3)=V_c + \pi R^2 h_{\mathrm{cy}}.$$

Substituindo:

$$
V(h_3)=0.03351 + \pi (0.20)^2 (0.70)
      =0.03351 + \pi(0.04)(0.70)
      =0.03351 + 0.028\pi.
$$

$$\boxed{V(1.50)\approx 0.1215 \text{ m}^3.}$$

Resumo dos resultados


$$
\begin{aligned}
V(0.40) &\approx 0.00421\ \text{m}^3, \\
V(0.80) &= 0.03351\ \text{m}^3, \\
V(1.50) &\approx 0.1215\ \text{m}^3.
\end{aligned}
$$

Esses cálculos são adequados para demonstração no quadro (caneta e papel), pois permitem ilustrar separadamente os regimes geométricos do hidrociclone.

__Código Python__

```python
import numpy as np

def volume_hidrociclone(h, R, Hc, Hcy):
    """
    Calcula o volume preenchido no hidrociclone para uma altura h.
    
    Parâmetros:
    h   -> altura do fluido medida a partir da ponta do cone (float)
    R   -> raio da base (float)
    Hc  -> altura da parte cônica (float)
    Hcy -> altura da parte cilíndrica (float)
    """
    
    # Volume total do cone
    Vc = (1/3) * np.pi * R**2 * Hc
    
    # Volume máximo total
    Vmax = Vc + np.pi * R**2 * Hcy
    
    # Caso 1: fluido apenas no cone
    if h <= 0:
        return 0.0
    
    elif h <= Hc:
        V = (np.pi * R**2 / (3 * Hc**2)) * h**3
        return V
    
    # Caso 2: cone cheio + cilindro parcialmente cheio
    elif h <= Hc + Hcy:
        V = Vc + np.pi * R**2 * (h - Hc)
        return V
    
    # Caso 3: acima do topo → volume saturado
    else:
        return Vmax
```

```python
# Exemplo de uso:
R   = 0.2   # raio (m)
Hc  = 0.4   # altura do cone (m)
Hcy = 0.6   # altura do cilindro (m)

for h in [0.0, 0.1, 0.4, 0.7, 1.0, 1.2]:
    V = volume_hidrociclone(h, R, Hc, Hcy)
    print(f"h = {h:.2f} m  ->  Volume = {V:.6f} m³")
```

Podemos refazer os cálculos para gerar um gráfico 

```python
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------------------------------
# Função que calcula o volume preenchido no hidrociclone
# ---------------------------------------------------------
def volume_hidrociclone(h, R, Hc, Hcy):
    """
    Calcula o volume preenchido no hidrociclone para uma altura h.
    
    Parâmetros:
    h   -> altura do fluido medida a partir da ponta do cone (float)
    R   -> raio da base (float)
    Hc  -> altura do cone (float)
    Hcy -> altura do cilindro (float)
    """
    
    # Volume total do cone
    Vc = (1/3) * np.pi * R**2 * Hc
    
    # Volume máximo (cone + cilindro)
    Vmax = Vc + np.pi * R**2 * Hcy
    
    # Caso 1: fluido ainda dentro do cone
    if h <= 0:
        return 0.0
    
    elif h <= Hc:
        return (np.pi * R**2 / (3 * Hc**2)) * h**3
    
    # Caso 2: cone cheio + cilindro parcialmente cheio
    elif h <= Hc + Hcy:
        return Vc + np.pi * R**2 * (h - Hc)
    
    # Caso 3: acima do topo
    else:
        return Vmax
```

```python
# ---------------------------------------------------------
# Parâmetros geométricos do hidrociclone
# ---------------------------------------------------------
R   = 0.2   # raio (m)
Hc  = 0.4   # altura do cone (m)
Hcy = 0.6   # altura do cilindro (m)

# ---------------------------------------------------------
# Geração dos dados do gráfico
# ---------------------------------------------------------
h_values = np.linspace(0, Hc + Hcy + 0.2, 300)
V_values = np.array([volume_hidrociclone(h, R, Hc, Hcy) for h in h_values])

# ---------------------------------------------------------
# Gráfico Volume × Altura
# ---------------------------------------------------------
plt.figure(figsize=(8,5))
plt.plot(h_values, V_values)
plt.xlabel("Altura do fluido h (m)")
plt.ylabel("Volume preenchido V(h) (m³)")
plt.title("Volume do Hidrociclone em função da Altura do Fluido")
plt.grid(True)

plt.show()
```
