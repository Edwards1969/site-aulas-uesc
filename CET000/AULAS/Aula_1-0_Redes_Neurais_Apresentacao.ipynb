  <strong>Definições:</strong> Blocos de construção: neurônios.

Antes de entender como uma rede neural completa funciona, precisamos começar pelo seu menor componente: o __neurônio__. Ele é a unidade básica de processamento de uma rede neural, inspirado no funcionamento dos __neurônios biológicos__, mas implementado de forma __matemática__ e __computacional__.

Um __neurônio artificial__ recebe um ou mais valores de entrada (chamados de __inputs__), realiza uma operação matemática com esses valores e gera um único valor de saída (__output__).

Abaixo uma figura que mostra a relação entre um __neurônio natural__ e uma __rede artificial__.

O funcionamento básico de um neurônio pode ser descrito em três etapas:

1. __Receber os dados de entrada__:
Cada entrada tem um valor numérico e está associada a um peso (que representa a importância dessa entrada).

3.  __Calcular uma combinação ponderada__:
O neurônio soma os produtos das entradas com seus respectivos pesos e adiciona um termo extra chamado bias (um tipo de ajuste fino).

3. __Aplicar uma função de ativação__:
Essa função decide qual será a saída final do neurônio, transformando o resultado da soma em um valor que pode ser usado na próxima camada da rede.

Abaixo está a estrutura de um neurônio com duas entradas, que pode ser representado assim:

<img src="img/Fig_01.png" alt="UESC" style="width: 500px; display: block; margin: 0 auto 5px auto;">
<p style="text-align: center;"><strong>Figura: Neurônio recebendo entradas.</strong></p>

Estão acontecendo três coisas nesse modelo:

- Primeiro, cada entrada é multiplicada por um peso:
  - Na __primeira__ entrada: $w_1 \cdot x_1$ 
  - Na __segunda__ entrada: $w_2 \cdot x_2$
- Em seguida, todas as entradas ponderadas são somadas com um __bias__: $b$
  - $ \sum (w_1 \cdot x_1 + w_2 \cdot x_2) + b$
- Finalmente, a soma é passada por meio de uma __função de ativação__:
  - $f(.)$ é a função de ativação,
  - $y$ é a saída,
  - $y = f(w_1 \cdot x_1 + w_2 \cdot x_2 + b)$

__Qual o papel do bias__:

A __função do bias__ (ou termo de viés) em uma rede neural __é deslocar a função de ativação para cima ou para baixo no eixo y__, o que dá mais flexibilidade ao modelo para ajustar-se aos dados.

- __Sem o bias__, a função de ativação __sempre passaria pela origem__ (quando todas as entradas forem zero, a saída também será zero).
- __Com o bias__, o neurônio pode __ajustar o limiar de ativação__ — ou seja, decidir quando ele "liga" ou "responde", mesmo que as entradas sejam todas zero.
- Isso é importante para aprender padrões mais complexos. O bias funciona como o "intercepto" de uma reta numa equação linear $y = mx + b$.


A __função de ativação__ é usada para transformar uma __entrada__ ilimitada em uma __saída__ que tem uma forma agradável e previsível. Uma função de ativação comumente usada é a __função sigmóide__ ($\sigma$):

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$
__Descrição__: Suaviza a entrada em uma curva em forma de S. Mapeia todos os valores para o intervalo $(0, 1)$.

<img src="img/Fig_02.png" alt="UESC" style="width: 400px; display: block; margin: 0 auto 5px auto;">
<p style="text-align: center;"><strong>Figura: Função Sigmoide.</strong></p>

```python
import numpy as np

x = 2
sigmoide = 1 / (1 + np.exp(-x))

print(f'Para x = {x} o valor de y = {sigmoide:.2f}' )
```

A função sigmoide gera apenas números no intervalo (1,0). Você pode pensar nisso como uma compactação ($\infty$, -$\infty$) - grandes números negativos tornam-se aproximadamente $0$ e grandes números positivos tornam-se aproximadamente $1$.

```python
import numpy as np
x1 = 6
x2 = -6
sigmoide1 = 1 / (1 + np.exp(-x1))
sigmoide2 = 1 / (1 + np.exp(-x2))

print(f'Para x = {x1} o valor de y = {sigmoide1:.2f}' )
print(f'Para x = {x2} o valor de y = {sigmoide2:.2f}' )
```

#### Função de ativação ReLU.

A função __ReLU__ (__Rectified Linear Unit__) retorna zero para valores negativos e retorna o próprio valor para valores positivos.

$$
\text{ReLU}(x) = \begin{cases}
0, & \text{se } x < 0 \\
x, & \text{se } x \geq 0
\end{cases}
$$
__Descrição__: zera valores negativos e mantém valores positivos.

A __ReLU__ é uma função de ativação usada em redes neurais que “__desliga__” o neurônio quando a entrada é __negativa__ e o “__liga__” proporcionalmente quando a entrada é __positiva__.

__E porque essa função é útil?__

- Porque é simples de calcular (não exige exponenciais como a _sigmoid_).
- Ajuda a evitar o problema do "desvanecimento do gradiente".
- Funciona bem em redes profundas (_deep learning_).

__Exemplo intuitivo__:

Imagine um neurônio com soma ponderada igual a zero. Se aplicarmos uma função como $ReLU = max(0, x)$, a saída será zero. Agora, se o bias for positivo, o resultado da soma pode ser positivo mesmo com entradas pequenas, ativando o neurônio.

> O bias permite que o neurônio modele dados mais complexos, deslocando a função de ativação e aumentando a capacidade de ajuste da rede.

#### Função de Ativação ReLU com Bias}

A função __ReLU__ com __bias__ é definida como:

$$
\text{ReLU}(x) = \max(0, w \cdot x + b)
$$

Ou, em forma de definição por partes:

$$
\text{ReLU}(x) =
\begin{cases}
0, & \text{se } w \cdot x + b < 0 \\
w \cdot x + b, & \text{se } w \cdot x + b \geq 0
\end{cases}
$$


Onde:

- valor de entrada do neurônio,
- $w$: peso (weight),
- $b$: bias (viés).


A __ReLU__ com __bias__ é aplicada após a combinação linear da entrada com os pesos e bias. Ela mantém a linearidade apenas para saídas positivas e zera as negativas, contribuindo para a não-linearidade do modelo e evitando saturação como ocorre na sigmoid.


#### Função de ativação tanh (Tangente hiperbólica)

$$
\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$
__Descrição__: Parecida com a sigmoid, mas mapeia para o intervalo $(-1, 1)$.

> Enquanto a função Sigmoid mapeia no intervalo de 0 a 1 a função de ativação Tanh mapeia no intervalo de $(-1, 1)$.

__Função de Ativação Softmax__

A função Softmax transforma um vetor de valores reais em um vetor de probabilidades. É usada principalmente na camada de saída de classificadores para problemas com múltiplas classes.

Definição matemática:

Seja $x = [x_1, x_2, ..., x_n]$ um vetor de entrada, então:

$$ Softmax(x_i)  = \frac{e^{x_i}}{\sum_{j=1}^n e^{x_j}}$$

__Características__:

- O resultado é um vetor de valores entre 0 e 1, e soma 1.
- Destaque os valores maiores do vetor (potencial para representar a classe mais provável).
- Bastante usada na saída de classificadores com múltiplas classes.

Vamos fazer uma comparação visual entre as funções __Sigmoid__, __ReLU__ e __Tanh__.

```python
import numpy as np
import matplotlib.pyplot as plt

# Entradas
x = np.linspace(-5, 5, 400)

# Funções de ativação
relu = lambda x: np.maximum(0, x)
sigmoid = lambda x: 1 / (1 + np.exp(-x))
tanh = lambda x: np.tanh(x)

# Aplicando as funções
y_relu = relu(x)
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)

# Gráfico comparativo
plt.figure(figsize=(10, 6))
plt.plot(x, y_relu, label='ReLU', linewidth=2, color='green')
plt.plot(x, y_sigmoid, label='Sigmoid', linewidth=2, color='blue')
plt.plot(x, y_tanh, label='tanh', linewidth=2, color='red')
plt.title("Comparação entre funções de ativação")
plt.xlabel("Entrada (x)")
plt.ylabel("Saída f(x)")
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

__Código com gráfico da softmax (para vetores)__

Como não podemos aplicar softmax diretamente sobre cada valor de x (como fizemos com sigmoid, tanh, etc.), o melhor é mostrar o efeito da softmax sobre vetores. Então faremos:

1. Mostrar o comportamento da softmax em vetores 1D.
2. Adicionar a curva ao gráfico comparativo para fins didáticos (mesmo sendo vetorial).

Aqui está a versão estendida do  código anterior incluído o softmax:


```python
import numpy as np
import matplotlib.pyplot as plt

# Entradas
x = np.linspace(-5, 5, 400)

# Funções de ativação
relu = lambda x: np.maximum(0, x)
sigmoid = lambda x: 1 / (1 + np.exp(-x))
tanh = lambda x: np.tanh(x)

# Função softmax para vetor de entrada
def softmax(x):
    exp_x = np.exp(x - np.max(x))  # para estabilidade numérica
    return exp_x / np.sum(exp_x)

# Aplicando as funções
y_relu = relu(x)
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)

# Para representar softmax, criamos um vetor de amostra
x_soft = np.array([1.0, 2.0, 3.0, 4.0])
y_soft = softmax(x_soft)

# Gráfico comparativo das funções escalares
plt.figure(figsize=(10, 6))
plt.plot(x, y_relu, label='ReLU', linewidth=2, color='green')
plt.plot(x, y_sigmoid, label='Sigmoid', linewidth=2, color='blue')
plt.plot(x, y_tanh, label='tanh', linewidth=2, color='red')
plt.title("Comparação entre funções de ativação (ReLU, Sigmoid, tanh)")
plt.xlabel("Entrada (x)")
plt.ylabel("Saída f(x)")
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()

# Gráfico separado para mostrar o efeito da softmax
plt.figure(figsize=(6, 4))
plt.bar(range(len(y_soft)), y_soft, color='purple')
plt.title("Saída da função Softmax para o vetor [1.0, 2.0, 3.0, 4.0]")
plt.xlabel("Índice do vetor")
plt.ylabel("Probabilidade (softmax)")
plt.grid(True, axis='y')
plt.show()
```

Vamos construir um exemplo simples em Python para mostrar a diferença entre um __neurônio com e sem bias__, usando uma função de __ativação ReLU__ (__função retificadora linear__). O objetivo é mostrar como o bias afeta a ativação do neurônio mesmo quando as entradas são pequenas ou até nulas.

__Código: Neurônio com e sem Bias__

```python
import numpy as np
import matplotlib.pyplot as plt

# Função de ativação ReLU
def relu(x):
    return np.maximum(0, x)

# Entradas (por simplicidade, usamos uma entrada apenas)
x = np.linspace(-2, 2, 400)

# Peso fixo
w = 1.0

# Sem bias
y_sem_bias = relu(w * x)

# Com bias
bias = 1.0
y_com_bias = relu(w * x + bias)

# Gráfico comparativo
plt.figure(figsize=(8, 5))
plt.plot(x, y_sem_bias, label="Sem bias", linestyle='--', color='blue')
plt.plot(x, y_com_bias, label="Com bias = +1", color='green')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("Efeito do Bias na Ativação de um Neurônio (ReLU)")
plt.xlabel("Entrada x")
plt.ylabel("Saída do neurônio")
plt.legend()
plt.grid(True)
plt.show()
```
