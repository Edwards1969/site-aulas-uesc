__Função__

Uma __função__ $f$ é uma regra que associa cada elemento $x \in A$ (onde $A$ é o __domínio__ a um único elemento $f(x) \in B$ (onde $B$ é o __contradomínio__. Escrevemos isso como:

$$f: A \rightarrow B$$

Em `Python`, funções são definidas com a palavra-chave `def`. Entre parênteses colocamos os __argumentos__ (se existirem). Normalmente, usamos o comando `return` para devolver uma saída.

Outra forma de criar funções em uma única linha é utilizando  `lambda`, também chamadas de __funções anônimas__.

__Método 1:__ 

A função $f(x) = 2x + 1$ pode ser escrita tanto com `def` quanto com `lambda`.  

**Exemplo 1:**

$f(x) = 2x + 1$

```python
# Usando def
def f(x):
    return 2*x + 1

# Usando lambda
f = lambda x: 2*x + 1

__Método 2:__ 

A função $G(x,y) = x + y$ também pode ser definida das duas formas.

**Exemplo 2:**

$G(x,y) = x + y$

````Python
# Usando def
def G(x, y):
    return x + y

# Usando lambda
G = lambda x, y: x + y
````

__A definição de convergência para sequência ($\epsilon, N$)__

Diz-se que uma sequência de números reais ($x_n$)  converge para $x \in R$ (e escrevemos $lim \ x = x$) se, para qualquer $\epsilon > 0$, existe um inteiro $N \in \mathbb{N}$  tal que, para todo $n > N, |x_n - x| < \epsilon$.

Considere a sequência definida pela seguinte expressão para cada termo:

a) $x_n = 1 - \frac{1}{n}$

b) $y_n = \frac{sin  \ n}{n}$

c) $z_n = \frac{ln \ n}{n}$

d) $E_n = (1 + \frac{1}{n})^n$

Para cada sequência, encontre o limite quando $n \rightarrow \infty$. Investigar a relação entre $\epsilon$  e $N$.

__Prova ($\varepsilon)-(N)$ (esboço):__ 

a) $x_n = 1 - \frac{1}{n}$

$
|x_n - 1| = \left|1 - \frac{1}{n} - 1\right| = \frac{1}{n}.
$

Para garantir ($\frac{1}{n}<\varepsilon$) basta escolher

$N=\left\lceil \frac{1}{\varepsilon}\right\rceil$


**(b) $y_n = \dfrac{\sin n}{n}$**

__Limite:__ $\lim_{n\to\infty} y_n = 0$.  

__Prova ($\varepsilon$-$N$) (esboço):__  

Como $|\sin n|\le 1$,  

$
|y_n| = \left|\frac{\sin n}{n}\right| \le \frac{1}{n}.
$

Assim, basta exigir $\frac{1}{n}<\varepsilon$, escolhendo  

$N=\left\lceil \frac{1}{\varepsilon}\right\rceil$  

**(c) $z_n = \dfrac{\ln n}{n}$**

__Limite:__ $\lim_{n\to\infty} z_n = 0$.  

__Prova ($\varepsilon$-$N$) (esboço):__  

Para $n$ suficientemente grande, vale a desigualdade $\ln n \le \sqrt{n}$. Logo,  

$
\frac{\ln n}{n} \le \frac{1}{\sqrt{n}}.
$

Para garantir $\dfrac{\ln n}{n}<\varepsilon$, basta tomar  

$N=\left\lceil \frac{1}{\varepsilon^{2}} \right\rceil$  

**(d) $E_n = \left(1 + \dfrac{1}{n}\right)^n$**

__Limite:__ $\lim_{n\to\infty} E_n = e$.  

__Prova ($\varepsilon$-$N$) (esboço):__  

Pela desigualdade clássica:  

$
\left(1+\frac{1}{n}\right)^n < e < \left(1+\frac{1}{n}\right)^{n+1},
$

temos  

$
0 < e - \left(1+\frac{1}{n}\right)^n < \frac{e}{n}.
$

Assim, para garantir $|E_n - e| < \varepsilon$, basta tomar  

$N=\left\lceil \frac{e}{\varepsilon} \right\rceil$

__Verificação numérica da convergência em Python__

Podemos calcular os primeiros termos de cada sequência e verificar como se aproximam do limite.

`````python
import numpy as np

# Definindo n grande
n = np.arange(1, 101)

# Sequências
x_n = 1 - 1/n
y_n = np.sin(n)/n
z_n = np.log(n)/n
E_n = (1 + 1/n)**n

# Mostrando os últimos 10 valores para ver a aproximação do limite
print("Últimos 10 valores de x_n:", x_n[-10:])
print("Últimos 10 valores de y_n:", y_n[-10:])
print("Últimos 10 valores de z_n:", z_n[-10:])
print("Últimos 10 valores de E_n:", E_n[-10:])
`````

__Intuição da definição de convergência__

Esta é uma das definições mais importantes em análise matemática. Ela fornece uma forma **rigorosa** de expressar o conceito intuitivo de convergência quando $n \to \infty$.

Sejam $x_n$ os termos de uma sequência e $x$ seu limite. Definimos o **erro** como

$$
\text{erro} = |x_n - x|
$$

ou seja, a diferença absoluta entre o $n$-ésimo termo e o limite.   

A definição de convergência diz que, à medida que avançamos na sequência, esse erro pode se tornar **arbitrariamente pequeno**.  

Em outras palavras, se alguém nos dá um $\varepsilon>0$ (tipicamente um número positivo muito pequeno), precisamos encontrar um ponto na sequência a partir do qual todos os termos seguintes satisfazem

$$
|x_n - x| < \varepsilon.
$$

O **índice $N$** em que isso acontece depende de $\varepsilon$: quanto menor for $\varepsilon$, maior será $N$. 


Exemplo: $x_n = 1$ - $\frac{1}{n}$

1. **Estimativa do limite**:  

   À medida que $n \to \infty$, o termo $1/n$ se torna desprezível, então podemos **chutar** que o limite é $x = 1$.

2. **Prova formal usando $\varepsilon$-$N$**: 
 

   O erro é
   $$
   |x_n - 1| = \frac{1}{n}.
   $$

   Para garantir $|x_n - 1| < \varepsilon$, basta escolher
   $$
   N = \left\lceil \frac{1}{\varepsilon} \right\rceil.
   $$

   Por exemplo, se $\varepsilon = 0.3$, qualquer inteiro $N \ge 10/3$ funciona (ex.: $N=4$).

3. **Observação**:  

Qualquer valor maior que $N$ também satisfaz a definição, mas é interessante pensar no **menor $N$** (denotado $N_{\min}$) que garante a convergência. É possível criar uma tabela de $N_{\min}$ para diferentes valores de $\varepsilon$ e visualizar como a sequência se aproxima do limite.


Tabela de $N_{\min}$ para $x_n = 1 - 1/n$

O menor $N$ que garante $|x_n - 1| < \varepsilon$ é dado por

$$
N_{\min} = \left\lceil \frac{1}{\varepsilon} \right\rceil
$$

Podemos calcular para diferentes valores de $\varepsilon$:

Veja a tabela abaixo:

__Tabela de $N_{\min}$ para $x_n = 1 - 1/n$__

O menor $N$ que garante $|x_n - 1| < \varepsilon$ é dado por:

$$
N_{\min} = \left\lceil \frac{1}{\varepsilon} \right\rceil
$$

| $\varepsilon$ | $N_{\min}$ |
|---------------|------------|
| 0.3           | 4          |
| 0.2           | 5          |
| 0.1           | 10         |
| 0.05          | 20         |
| 0.03          | 34         |


`````python
import math

# Valores de epsilon
epsilons = [0.3, 0.2, 0.1, 0.05, 0.03]

# Calculando N_min
N_min = [math.ceil(1/eps) for eps in epsilons]

# Mostrando a tabela
print("ε      | N_min")
print("-------|-------")
for eps, N in zip(epsilons, N_min):
    print(f"{eps:<6} | {N}")
`````

Cálculo usando programação em Python do problema acima.

`````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

def plot_convergence(n_min=1, n_max=40, logscale=False, marker_every=1,
                     figsize=(6,4), ylim=None, xlim=None):
    """
    Plota |x_n - 1|, |y_n|, |z_n|, |E_n - e| para n em [n_min, n_max].
    """
    n = np.arange(n_min, n_max + 1)
    recp = 1/n

    # Sequências e erros + estilos + labels
    erros = [
        (np.abs((1 - recp) - 1),   'o-',  r'$|x_n - 1|$', 'C0'),
        (np.abs(np.sin(n)/n),      's--', r'$|y_n|$',     'C1'),
        (np.abs(np.log(n)/n),      'D-.', r'$|z_n|$',     'C2'),
        (np.abs((1 + recp)**n - np.e), '^:', r'$|E_n - e|$', 'C3')
    ]

    plt.figure(figsize=figsize)
    legend_handles = []

    for erro, style, label, color in erros:
        marker = style[0]; linestyle = style[1:]
        # plota linha real
        plt.plot(n, erro, style, color=color, linewidth=1.0,
                 markersize=4, markevery=marker_every)
        # cria legenda com mesma cor, marcador e linha
        legend_handles.append(Line2D([0], [0], linestyle=linestyle, marker=marker,
                                     color=color, markersize=5, linewidth=1.2, label=label))

    plt.xlabel(r'$n$', fontsize=12)
    plt.ylabel(r'$|Erro|$', fontsize=12)
    plt.title('Convergência das sequências', fontsize=14)

    if logscale: plt.yscale('log')
    if xlim: plt.xlim(xlim)
    else: plt.xlim(n_min-1, n_max+1)
    if ylim: plt.ylim(ylim)

    plt.grid(True, which="both", ls="--", lw=0.5)
    plt.legend(handles=legend_handles, loc='upper right', fontsize=10, framealpha=0.9)
    plt.tight_layout()
    plt.show()

# Exemplo
plot_convergence(n_min=1, n_max=40, logscale=False, marker_every=1, ylim=(0,0.4), figsize=(6,4))
`````

Cálculo abaixo esta no livro: Chongchitnan, Siri. "Exploring University Mathematics with Python" Editora: Springer

`````python
# Código da fonte: https://github.com/siriwarwick/book

import numpy as np
import matplotlib.pyplot as plt

n = np.arange(1,41)

recp = 1/n
xn = 1-recp
err_xn = recp

yn = np.sin(n)/n
err_yn = abs(yn)

zn = np.log(n)/n
err_zn = abs(zn)

En = (1 + recp)**n
err_En = abs(En-np.e)

plt.plot(n, err_xn, 'o-', ms=3)
plt.plot(n, err_yn, 'o--', ms=3)
plt.plot(n, err_zn, 'o-', ms=3)
plt.plot(n, err_En, 'o:', ms=3)

plt.xlim(0,40)
plt.ylim(0,0.4)
plt.ylabel(r'$n$')
plt.xlabel('|Error|')
plt.legend([r'$|z_n-1|$', r'$|y_n|$', r'$|z_n|$', r'$|z_n|$', r'$|E_n-e|$'])
plt.grid('on')
plt.show()
`````
